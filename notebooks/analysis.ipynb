```python
import pandas as pd
from pathlib import Path

# Die CSV-Datei liegt im Ordner "data" relativ zum Notebook.
# Falls das Notebook aus einem Unterordner gestartet wird, wird eine Ebene
# nach oben navigiert, damit der Pfad trotzdem stimmt.
csv_path = Path.cwd() / "data" / "dbo.vw_batch_kpis_converted.csv"
if not csv_path.exists():
    csv_path = Path.cwd().parent / "data" / "dbo.vw_batch_kpis_converted.csv"

# Die Datei ist semicolon-getrennt und verwendet das deutsche Dezimalkomma.
# utf-8-sig entfernt automatisch die BOM (Byte Order Mark), die Excel beim
# Export manchmal an den Dateianfang schreibt.
df = pd.read_csv(
    csv_path,
    sep=";",
    decimal=",",
    encoding="utf-8-sig",
    low_memory=False,
)

print("Shape:", df.shape)
df.head(20)
```

```python
import numpy as np
import warnings
warnings.filterwarnings("ignore")

# Beide Cobas8000-Geräte teilen sich denselben line_name in der Datenbank.
# Um gezielt nur ein Gerät auszuwählen, filtert man zusätzlich über die
# eindeutige line_oid. LINE_LABEL wird nur für Diagrammtitel verwendet.
LINE       = "Cobas8000-"
LINE_OID   = "C5CEB0CA-FC9E-4979-A2DB-6A77B6F330A7"  # entspricht Cobas8000-1
LINE_LABEL = "Cobas8000-1"

# Nur Datensätze der gewählten Linie und des gewählten Geräts behalten.
df_line = df[(df["line_name"] == LINE) & (df["line_oid"] == LINE_OID)].copy()

# OEE und Mengenspalte in ihre korrekten Datentypen umwandeln.
df_line["oee"]            = pd.to_numeric(df_line["oee"],            errors="coerce")
df_line["total_quantity"] = pd.to_numeric(df_line["total_quantity"], errors="coerce")

# Startdatum und -uhrzeit aus zwei separaten CSV-Spalten zusammenführen:
#   "start"       enthält nur das Datum (z. B. "2024-07-15")
#   "Unnamed: 14" enthält nur die Uhrzeit (z. B. "08:23:00")
# Erst der kombinierte Wert ergibt einen vollständigen Startzeitpunkt.
df_line["start"] = pd.to_datetime(
    df_line["start"].astype(str) + " " + df_line["Unnamed: 14"].astype(str),
    errors="coerce",
)
# Stoppzeitpunkt analog zusammenführen ("stop" + "Unnamed: 16").
df_line["stop_dt"] = pd.to_datetime(
    df_line["stop"].astype(str) + " " + df_line["Unnamed: 16"].astype(str),
    errors="coerce",
)

# Zeilen ohne OEE-Wert, Startzeit oder Stoppzeit sind nicht nutzbar.
df_line = df_line.dropna(subset=["oee", "start", "stop_dt"])

# Chargen ohne jede Produktion explizit ausschließen.
n_zero = (df_line["total_quantity"] == 0).sum()
df_line = df_line[df_line["total_quantity"] > 0].copy()
if n_zero > 0:
    print(f"  {n_zero} Charge(n) entfernt: keine Produktionsmenge vorhanden (Leerläufe).")

# Chargen ohne verwertbaren OEE entfernen (Totalausfall oder Datenfehler).
n_invalid = ((df_line["oee"] <= 0) | (df_line["oee"] > 1)).sum()
df_line = df_line[(df_line["oee"] > 0) & (df_line["oee"] <= 1)].copy()
if n_invalid > 0:
    print(f"  {n_invalid} Charge(n) entfernt: kein Output erzielt oder ungültiger Datenwert.")

# OEE von Dezimalzahl (0–1) auf Prozent umrechnen.
df_line["oee_pct"] = df_line["oee"] * 100

# Chargendauer in Minuten berechnen — Grundlage für die Zeitgewichtung.
# Eine längere Charge soll im Durchschnitt stärker gewichtet werden,
# weil sie mehr Betriebszeit repräsentiert als eine kurze Charge.
df_line["duration_min"] = (df_line["stop_dt"] - df_line["start"]).dt.total_seconds() / 60

# Chargen mit nicht auflösbarer Dauer ausschließen (Stop ≤ Start).
n_no_dur = (df_line["duration_min"] <= 0).sum()
df_line = df_line[df_line["duration_min"] > 0].copy()
if n_no_dur > 0:
    print(f"  {n_no_dur} Charge(n) entfernt: keine verwertbare Chargendauer (Start ≥ Stop).")

# Für den Zeitreihenverlauf aggregieren wir auf Monatsebene.
df_line["month"] = df_line["start"].dt.to_period("M")

# Zeitgewichteter monatlicher Mittelwert:
#   np.average(..., weights=duration_min) gewichtet jeden OEE-Wert
#   proportional zur Chargendauer — längere Chargen gehen stärker ein.
# Standardabweichung (σ) und Anzahl bleiben ungewichtet, da sie die
# Streuung der Chargen, nicht der Betriebszeit, beschreiben.
monthly = (
    df_line.groupby("month")
    .apply(lambda x: pd.Series({
        "mean":  np.average(x["oee_pct"], weights=x["duration_min"]),
        "std":   x["oee_pct"].std(),
        "count": len(x),
    }))
    .reset_index()
)
monthly["month_dt"] = monthly["month"].dt.to_timestamp()
monthly["std"] = monthly["std"].fillna(0)
# groupby().apply() speichert alle Spalten als float64 — count zurück auf int setzen,
# damit Hover-Tooltips "Chargen: 20" statt "Chargen: 20.0" zeigen.
monthly["count"] = monthly["count"].astype(int)

# Zeitgewichteter Gesamtmittelwert direkt aus den Einzelchargen.
# Monate mit wenigen langen Chargen werden nicht übergewichtet.
overall_mean = np.average(df_line["oee_pct"], weights=df_line["duration_min"])

print(f"Datenbasis {LINE_LABEL}: {len(df_line)} Chargen | {monthly['month'].min()} bis {monthly['month'].max()}")
print(f"  Entfernt: {n_zero} Leerläufe | {n_invalid} ungültige OEE-Werte | {n_no_dur} Chargen ohne Dauer")
print(f"Durchschnittlicher OEE (zeitgewichtet): {overall_mean:.1f} %")


print()
```

# OEE-Werte Cobas8000-1: Bibliotheksvergleich

Alle drei Diagramme zeigen **dieselbe Darstellung**: monatlicher Mittelwert-OEE der Produktionslinie  
**Cobas8000-1**  + Gesamtmittelwert-Referenzlinie + ±1σ-Band.

| Eigenschaft | **Matplotlib** | **Seaborn** | **Plotly** |
|---|---|---|---|
| Stil-Ansatz | Manuell / Low-Level | Theme-basiert / deklarativ | Interaktiv / webbasiert |
| Fehlerband | `ax.fill_between` manuell | `errorbar="sd"` in einem Aufruf | `go.Scatter` fill-to-next |
| Einzelpunkte | `ax.scatter` | `stripplot` | `go.Scatter` overlay |
| Interaktivität | Keine (statisch) | Keine (statisch) | Hover, Zoom, Pan |
| Ausgabeformat | PNG/SVG | PNG/SVG | HTML (inline im Notebook) |


## 1 · Matplotlib — vollständige manuelle Kontrolle

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# Matplotlib arbeitet auf sehr niedriger Ebene: jedes Element des Diagramms
# (Fläche, Linie, Punkte, Beschriftungen) wird einzeln und explizit gezeichnet.
# Das gibt maximale Kontrolle, erfordert aber auch mehr Code.

# Hintergrund des gesamten Figure-Objekts
# Hintergrund der Zeichenfläche selbst

fig, ax = plt.subplots(figsize=(12, 5))
fig.patch.set_facecolor("#f8f8f8")   
ax.set_facecolor("#f0f0f0")          

# Hilfsvariablen für bessere Lesbarkeit

# Zeitachse (Monatsstempel)
# monatlicher OEE-Mittelwert
# monatliche Standardabweichung

x = monthly["month_dt"]   
y = monthly["mean"]       
s = monthly["std"]        

# Das ±1σ-Band visualisiert die Streuung innerhalb eines Monats.
# fill_between füllt den Bereich zwischen (y - s) und (y + s) mit einer
# halbtransparenten Fläche (alpha=0.18) in der Hauptfarbe.
ax.fill_between(x, y - s, y + s, color="#4C72B0", alpha=0.18, label="±1 Std.-Abweichung")

# Die Mittelwertlinie liegt optisch über dem Band (höherer zorder-Standard).
# Die hohlen Marker (weiß gefüllt) heben jeden Monatswert hervor.
ax.plot(x, y, color="#4C72B0", linewidth=2.2, marker="o", markersize=6,
        markerfacecolor="white", markeredgewidth=1.8, label="Monatl. Ø OEE")

# Die Einzelchargen werden als sehr kleine, fast transparente Punkte
# im Hintergrund (zorder=1) gezeichnet, um die Datengrundlage sichtbar zu machen.
batch_x = df_line["start"]
batch_y = df_line["oee_pct"]
ax.scatter(batch_x, batch_y, color="#4C72B0", alpha=0.08, s=8, zorder=1, label="Einzelchargen")

# Die horizontale Referenzlinie zeigt den chargenseitig gewichteten Gesamtmittelwert.
ax.axhline(overall_mean, color="#C44E52", linewidth=1.8, linestyle="--",
           label=f"Gesamtmittelwert: {overall_mean:.1f} %")

# Jeden Monatspunkt mit dem gerundeten OEE-Wert beschriften.
# xytext=(0, 7) verschiebt die Beschriftung 7 Punkte nach oben.
for _, row in monthly.iterrows():
    ax.annotate(f"{row['mean']:.0f}", xy=(row["month_dt"], row["mean"]),
                xytext=(0, 7), textcoords="offset points",
                ha="center", fontsize=7, color="#333333")

# X-Achse: jeden Monat als Tick, Format "Jan\n2025"
ax.xaxis.set_major_locator(mdates.MonthLocator())
ax.xaxis.set_major_formatter(mdates.DateFormatter("%b\n%Y"))
ax.set_xlabel("Monat", fontsize=11)
ax.set_ylabel("OEE [%]", fontsize=11)
ax.set_title(
    f"Monatlicher OEE-Verlauf — {LINE_LABEL} (Matplotlib)\n"
    f"{n_zero} Leerläufe und {n_invalid} Chargen ohne verwertbaren OEE entfernt "
    f"| {len(df_line)} Chargen in der Auswertung",
    fontsize=13, fontweight="bold")
ax.set_ylim(0, 100)
ax.legend(fontsize=9, loc="lower left")
ax.grid(axis="y", linestyle="--", alpha=0.4)

# Obere und rechte Rahmenlinie entfernen – wirkt aufgeräumter.
for spine in ["top", "right"]:
    ax.spines[spine].set_visible(False)


plt.tight_layout()
plt.show()
```

## 2 · Seaborn — Theme-basiert, deklarativ

```python
import seaborn as sns

# Seaborn setzt ein globales Theme, das alle folgenden matplotlib-Plots
# beeinflusst. "whitegrid" gibt ein sauberes weißes Raster, "muted" dämpft
# die Farben etwas ab. Am Ende der Zelle setzen wir die Defaults zurück.
sns.set_theme(style="whitegrid", palette="muted", font_scale=1.05)

# Seaborn erwartet "long-format" Daten: eine Zeile pro Beobachtung.
# Wir konvertieren die Period-Spalte in einen lesbaren String, weil Seaborn
# kategoriale X-Achsen besser mit Strings als mit Period-Objekten umgehen kann.
df_plot = df_line.copy()
df_plot["month_str"] = df_plot["month"].dt.strftime("%Y-%m")

# Die Monate müssen explizit sortiert übergeben werden, sonst ordnet Seaborn
# die Kategorien alphabetisch – was hier zufällig stimmt, aber nicht garantiert ist.
month_order = sorted(df_plot["month_str"].unique())

fig, ax = plt.subplots(figsize=(12, 5))

# stripplot streut die Einzelchargen als kleine, halbtransparente Punkte
# leicht zufällig (jitter=True) um die Monatskategorie herum, damit
# übereinanderliegende Punkte sichtbar bleiben.
sns.stripplot(
    data=df_plot, x="month_str", y="oee_pct",
    order=month_order,
    color="#4C72B0", alpha=0.12, size=3, jitter=True, ax=ax, zorder=1,
)

# pointplot berechnet Mittelwert und Fehlerbalken intern.
# errorbar="sd" zeichnet ±1 Standardabweichung als Fehlerbalken mit Caps.
# Das ersetzt in Seaborn die manuelle fill_between-Logik aus Matplotlib.
sns.pointplot(
    data=df_plot, x="month_str", y="oee_pct",
    order=month_order,
    errorbar="sd",
    color="#4C72B0",
    markers="o",
    linestyles="-",
    err_kws={"linewidth": 1.5, "alpha": 0.6},
    capsize=0.12,
    ax=ax,
    zorder=3,
    label="Monatl. Ø OEE ± 1σ",
)

# Referenzlinie für den Gesamtdurchschnitt – identisch zu Matplotlib,
# da Seaborn intern ebenfalls matplotlib-Axes verwendet.
ax.axhline(overall_mean, color="#c44e52", linewidth=1.8, linestyle="--",
           label=f"Gesamtmittelwert: {overall_mean:.1f} %", zorder=4)

# Da die X-Achse kategorial ist, müssen die Labels manuell gesetzt werden.
ax.set_xticklabels(month_order, rotation=45, ha="right", fontsize=8)
ax.set_xlabel("Monat")
ax.set_ylabel("OEE [%]")
ax.set_ylim(0, 100)
ax.set_title(
    f"Monatlicher OEE-Verlauf — {LINE_LABEL} (Seaborn)\n"
    f"{n_zero} Leerläufe und {n_invalid} Chargen ohne verwertbaren OEE entfernt "
    f"| {len(df_line)} Chargen in der Auswertung",
    fontsize=13, fontweight="bold")
ax.legend(fontsize=9, loc="lower left")

plt.tight_layout()
plt.show()

# Theme zurücksetzen, damit nachfolgende Matplotlib-Zellen nicht beeinflusst werden.

sns.reset_defaults()
```

## 3 · Plotly — interaktiv, webbasiert

```python
import plotly.graph_objects as go

# Plotly erzeugt interaktive HTML-Diagramme, die direkt im Notebook rendern.
# Jedes visuelle Element (Scatter, Band, Linie) ist ein eigener "Trace",
# der dem Figure-Objekt hinzugefügt wird. Reihenfolge = Zeichenreihenfolge.

# Hilfsvariablen für die monatlichen Werte
x      = monthly["month_dt"]
y_mean = monthly["mean"]
y_std  = monthly["std"]
y_up   = y_mean + y_std   # obere Bandgrenze
y_lo   = y_mean - y_std   # untere Bandgrenze

fig = go.Figure()

# Trace 1: Einzelchargen im Hintergrund.
# opacity=0.15 macht sie kaum sichtbar, aber beim Hovern über einen Punkt
# erscheinen Datum und exakter OEE-Wert als Tooltip.
fig.add_trace(go.Scatter(
    x=df_line["start"],
    y=df_line["oee_pct"],
    mode="markers",
    name="Einzelchargen",
    marker=dict(color="#4C72B0", size=4, opacity=0.15),
    hovertemplate="Datum: %{x|%Y-%m-%d}<br>OEE: %{y:.1f} %<extra>Einzelcharge</extra>",
))

# Trace 2: ±1σ-Band als geschlossene Fläche.
# Plotly kennt kein fill_between, daher wird der Pfad manuell konstruiert:
# erst von links nach rechts entlang der oberen Grenze, dann rückwärts
# entlang der unteren Grenze – das ergibt eine geschlossene Kurve.
# fill="toself" füllt diese Fläche. Die Linie selbst ist unsichtbar (alpha=0).
fig.add_trace(go.Scatter(
    x=list(x) + list(x[::-1]),
    y=list(y_up) + list(y_lo[::-1]),
    fill="toself",
    fillcolor="rgba(76, 114, 176, 0.18)",
    line=dict(color="rgba(255,255,255,0)"),
    hoverinfo="skip",       # kein Tooltip für die Bandfläche selbst
    showlegend=True,
    name="±1 Std.-Abweichung",
))

# Trace 3: monatliche Mittelwertlinie.
# customdata bündelt Zusatzinfos (σ, Chargenzahl) pro Punkt, die im
# hovertemplate über %{customdata[i]} referenziert werden.
fig.add_trace(go.Scatter(
    x=x,
    y=y_mean,
    mode="lines+markers",
    name="Monatl. Ø OEE",
    line=dict(color="#4C72B0", width=2.5),
    marker=dict(size=8, color="white", line=dict(color="#4C72B0", width=2)),
    hovertemplate=(
        "<b>%{x|%B %Y}</b><br>"
        "Ø OEE: %{y:.1f} %<br>"
        "σ: %{customdata[0]:.1f} %<br>"
        "Chargen: %{customdata[1]}<extra></extra>"
    ),
    customdata=np.column_stack([y_std, monthly["count"]]),
))

# Horizontale Referenzlinie für den Gesamtmittelwert.
# add_hline ist eine Hilfsmethode, die intern ebenfalls einen Shape-Layer erzeugt.
fig.add_hline(
    y=overall_mean,
    line=dict(color="#C44E52", width=2, dash="dash"),
    annotation_text=f"Ø {overall_mean:.1f} %",
    annotation_position="top right",
    annotation_font=dict(color="#C44E52", size=12),
)

# Layout: Achsenbeschriftung, Gitter, Hintergrundfarben.
# hovermode="x unified" zeigt beim Hovern alle Traces an einem X-Wert
# gleichzeitig in einem gemeinsamen Tooltip.
fig.update_layout(
    title=dict(
        text=(
            f"Monatlicher OEE-Verlauf — {LINE_LABEL} (Plotly)<br>"
            f"<sup>{n_zero} Leerläufe und {n_invalid} Chargen ohne verwertbaren OEE entfernt "
            f"| {len(df_line)} Chargen in der Auswertung</sup>"
        ),
        font=dict(size=16),
    ),
    xaxis=dict(
        title="Monat",
        tickformat="%b\n%Y",
        dtick="M1",           # ein Tick pro Monat
        showgrid=True,
        gridcolor="#e0e0e0",
    ),
    yaxis=dict(
        title="OEE [%]",
        range=[0, 100],
        showgrid=True,
        gridcolor="#e0e0e0",
    ),
    legend=dict(orientation="h", yanchor="bottom", y=1.02, xanchor="right", x=1),
    plot_bgcolor="white",
    paper_bgcolor="#f8f8f8",
    width=1000,
    height=500,
    hovermode="x unified",
)


fig.show()
```

## 4 · Cobas8000-2 — Vergleich der Gegenlinie

```python
import plotly.graph_objects as go

LINE2       = "Cobas8000-"
LINE2_OID   = "074FADAA-59C9-420A-BE0E-6367E00BE429"
LINE2_LABEL = "Cobas8000-2"

# Filtern, Typen konvertieren, Start- und Stoppdatetime zusammenführen.
df2 = df[(df["line_name"] == LINE2) & (df["line_oid"] == LINE2_OID)].copy()
df2["oee"]            = pd.to_numeric(df2["oee"],            errors="coerce")
df2["total_quantity"] = pd.to_numeric(df2["total_quantity"], errors="coerce")

df2["start"] = pd.to_datetime(
    df2["start"].astype(str) + " " + df2["Unnamed: 14"].astype(str),
    errors="coerce",
)
df2["stop_dt"] = pd.to_datetime(
    df2["stop"].astype(str) + " " + df2["Unnamed: 16"].astype(str),
    errors="coerce",
)
df2 = df2.dropna(subset=["oee", "start", "stop_dt"])

# Chargen ohne jede Produktion explizit ausschließen.
n_zero2 = (df2["total_quantity"] == 0).sum()
df2 = df2[df2["total_quantity"] > 0].copy()
if n_zero2 > 0:
    print(f"  {n_zero2} Charge(n) entfernt: keine Produktionsmenge vorhanden (Leerläufe).")

# Ungültige OEE-Werte entfernen.
n_invalid2 = ((df2["oee"] <= 0) | (df2["oee"] > 1)).sum()
df2 = df2[(df2["oee"] > 0) & (df2["oee"] <= 1)].copy()
if n_invalid2 > 0:
    print(f"  {n_invalid2} Charge(n) entfernt: kein Output erzielt oder ungültiger Datenwert.")

df2["oee_pct"] = df2["oee"] * 100

# Chargendauer berechnen und Chargen ohne verwertbare Dauer ausschließen.
df2["duration_min"] = (df2["stop_dt"] - df2["start"]).dt.total_seconds() / 60
n_no_dur2 = (df2["duration_min"] <= 0).sum()
df2 = df2[df2["duration_min"] > 0].copy()
if n_no_dur2 > 0:
    print(f"  {n_no_dur2} Charge(n) entfernt: keine verwertbare Chargendauer (Start ≥ Stop).")

# Zeitgewichtete monatliche Aggregation
df2["month"] = df2["start"].dt.to_period("M")
monthly2 = (
    df2.groupby("month")
    .apply(lambda x: pd.Series({
        "mean":  np.average(x["oee_pct"], weights=x["duration_min"]),
        "std":   x["oee_pct"].std(),
        "count": len(x),
    }))
    .reset_index()
)
monthly2["month_dt"] = monthly2["month"].dt.to_timestamp()
monthly2["std"] = monthly2["std"].fillna(0)
# groupby().apply() speichert alle Spalten als float64 — count zurück auf int.
monthly2["count"] = monthly2["count"].astype(int)

# Zeitgewichteter Gesamtmittelwert
overall_mean2 = np.average(df2["oee_pct"], weights=df2["duration_min"])

print(f"Datenbasis {LINE2_LABEL}: {len(df2)} Chargen | {monthly2['month'].min()} bis {monthly2['month'].max()}")
print(f"  Entfernt: {n_zero2} Leerläufe | {n_invalid2} ungültige OEE-Werte | {n_no_dur2} Chargen ohne Dauer")
print(f"Durchschnittlicher OEE (zeitgewichtet): {overall_mean2:.1f} % (zum Vergleich Cobas8000-1: {overall_mean:.1f} %)")

# Hilfsvariablen für den Plot
x2     = monthly2["month_dt"]
y2     = monthly2["mean"]
y2_std = monthly2["std"]
y2_up  = y2 + y2_std
y2_lo  = y2 - y2_std

fig2 = go.Figure()

# Trace 1: Einzelchargen – orange statt blau, um Cobas8000-2 farblich
# von Cobas8000-1 zu unterscheiden.
fig2.add_trace(go.Scatter(
    x=df2["start"], y=df2["oee_pct"],
    mode="markers", name="Einzelchargen",
    marker=dict(color="#DD8452", size=4, opacity=0.15),
    hovertemplate="Datum: %{x|%Y-%m-%d}<br>OEE: %{y:.1f} %<extra>Einzelcharge</extra>",
))

# Trace 2: ±1σ-Band
fig2.add_trace(go.Scatter(
    x=list(x2) + list(x2[::-1]),
    y=list(y2_up) + list(y2_lo[::-1]),
    fill="toself",
    fillcolor="rgba(221, 132, 82, 0.18)",
    line=dict(color="rgba(255,255,255,0)"),
    hoverinfo="skip", showlegend=True,
    name="±1 Std.-Abweichung",
))

# Trace 3: zeitgewichtete monatliche Mittelwertlinie
fig2.add_trace(go.Scatter(
    x=x2, y=y2,
    mode="lines+markers", name="Monatl. Ø OEE (zeitgew.)",
    line=dict(color="#DD8452", width=2.5),
    marker=dict(size=8, color="white", line=dict(color="#DD8452", width=2)),
    hovertemplate=(
        "<b>%{x|%B %Y}</b><br>"
        "Ø OEE: %{y:.1f} %<br>"
        "σ: %{customdata[0]:.1f} %<br>"
        "Chargen: %{customdata[1]}<extra></extra>"
    ),
    customdata=np.column_stack([y2_std, monthly2["count"]]),
))

# Referenzlinie für den eigenen zeitgewichteten Gesamtmittelwert
fig2.add_hline(
    y=overall_mean2,
    line=dict(color="#DD8452", width=2, dash="dash"),
    annotation_text=f"Ø {LINE2_LABEL}: {overall_mean2:.1f} %",
    annotation_position="top left",
    annotation_font=dict(color="#DD8452", size=11),
)

# Zweite Referenzlinie für Cobas8000-1 als Vergleich
fig2.add_hline(
    y=overall_mean,
    line=dict(color="#4C72B0", width=1.5, dash="dot"),
    annotation_text=f"Ø Cobas8000-1: {overall_mean:.1f} %",
    annotation_position="top right",
    annotation_font=dict(color="#4C72B0", size=11),
)

fig2.update_layout(
    title=dict(
        text=(
            f"Monatlicher OEE-Verlauf — {LINE2_LABEL} (Plotly)<br>"
            f"<sup>{n_zero2} Leerläufe und {n_invalid2} Chargen ohne verwertbaren OEE entfernt "
            f"| {len(df2)} Chargen in der Auswertung | Vergleich: Cobas8000-1 "
            f"| Mittelwert zeitgewichtet nach Chargendauer</sup>"
        ),
        font=dict(size=16),
    ),
    xaxis=dict(
        title="Monat", tickformat="%b\n%Y", dtick="M1",
        showgrid=True, gridcolor="#e0e0e0",
    ),
    yaxis=dict(
        title="OEE [%]", range=[0, 100],
        showgrid=True, gridcolor="#e0e0e0",
    ),
    legend=dict(orientation="h", yanchor="bottom", y=1.02, xanchor="right", x=1),
    plot_bgcolor="white", paper_bgcolor="#f8f8f8",
    width=1000, height=500, hovermode="x unified",
)


fig2.show()
```

```python
import plotly.graph_objects as go

# ── 6 · Cobas8000-2 — rollierendes BDP-Potenzial (dynamisches 90. Perzentil) ─
#
# Statt einer einzigen statischen Linie über 24 Monate wird das 90. Perzentil
# quartalsweise neu berechnet. Das ergibt eine Treppenlinie, die zeigt, wie
# sich das erreichbare Potenzial der Anlage im Zeitverlauf verändert hat.
#
# Methodik:
#   - Perioden: Q1 2024, Q2 2024, … Q4 2025, Q1 2026
#   - Je Quartal: 90. Perzentil aller Chargen dieses Quartals
#   - Darstellung als Treppenstufe (step="post"): der Wert gilt ab Quartalsbeginn
#     bis das nächste Quartal beginnt.
#
# Warum Quartale und nicht Monate?
#   Ein monatliches Perzentil ist bei oft nur 20–40 Chargen pro Monat statistisch
#   instabil (ein einzelner Ausreißer bewegt das 90. Perz. stark). Quartale
#   bündeln ~60–120 Chargen und liefern robustere Schätzwerte.

# df2 und target_oee2 / overall_mean2 stammen aus der vorherigen Zelle.
# Sicherheitshalber explizit referenzieren:
_df = df2.copy()
_df["quarter"] = _df["start"].dt.to_period("Q")

# 90. Perzentil je Quartal berechnen.
# min_count=1 stellt sicher, dass Quartale mit nur wenigen Chargen nicht
# stillschweigend übersprungen werden.
bdp_quarterly = (
    _df.groupby("quarter")["oee_pct"]
    .quantile(0.90)
    .reset_index()
    .rename(columns={"oee_pct": "bdp_q90"})
)

# Quartal → Zeitstempel (Quartalsbeginn) für die X-Achse
bdp_quarterly["q_start"] = bdp_quarterly["quarter"].dt.to_timestamp("D", how="start")

# Quartal-Ende für die Treppenstufe: letzter Tag des Quartals als nächster
# Quartalsbeginn. So reicht jede Stufe exakt bis zum nächsten Quartalswechsel.
bdp_quarterly["q_end"] = bdp_quarterly["quarter"].dt.to_timestamp("D", how="end")

# Für eine saubere Treppenlinie in Plotly: jeden Wert zweimal eintragen –
# einmal am Anfang der Periode, einmal am Ende. Das ergibt exakt waagerechte
# Stufen ohne diagonale Verbindungslinien.
step_x, step_y = [], []
for _, row in bdp_quarterly.iterrows():
    step_x += [row["q_start"], row["q_end"]]
    step_y += [row["bdp_q90"], row["bdp_q90"]]

# Annotationstext je Quartal für den Hover-Tooltip
bdp_quarterly["label"] = bdp_quarterly.apply(
    lambda r: f"Q{r['quarter'].quarter} {r['quarter'].year}: {r['bdp_q90']:.1f} %", axis=1
)

print("Rollierendes BDP-Potenzial (90. Perzentil je Quartal):\n")
for _, row in bdp_quarterly.iterrows():
    n = int(_df[_df["quarter"] == row["quarter"]]["oee_pct"].count())
    print(f"  {row['label']}  ({n} Chargen)")

print(f"\n  Statisches Gesamt-BDP zum Vergleich: {target_oee2:.1f} %")
print(f"  Gesamtmittelwert Status Quo:          {overall_mean2:.1f} %")

# ─── Diagramm ────────────────────────────────────────────────────────────────
fig_rolling = go.Figure()

# Trace 1: Einzelchargen (Standard, orange blass)
fig_rolling.add_trace(go.Scatter(
    x=df2["start"], y=df2["oee_pct"],
    mode="markers", name="Einzelchargen",
    marker=dict(color="#DD8452", size=4, opacity=0.22),
    hovertemplate="Datum: %{x|%Y-%m-%d}<br>OEE: %{y:.1f} %<extra>Einzelcharge</extra>",
))

# Trace 2: Golden Batches (grün, leicht opaker)
mask_golden_r = df2["oee_pct"] >= target_oee2
fig_rolling.add_trace(go.Scatter(
    x=df2[mask_golden_r]["start"], y=df2[mask_golden_r]["oee_pct"],
    mode="markers", name=f"Golden Batch (≥ stat. BDP {target_oee2:.0f} %)",
    marker=dict(color="#2CA02C", size=4, opacity=0.45, symbol="circle"),
    hovertemplate=(
        "<b>Golden Batch</b><br>Datum: %{x|%Y-%m-%d}<br>OEE: %{y:.1f} %"
        "<extra>Golden Batch</extra>"
    ),
))

# Trace 3: monatliche Mittelwertlinie
fig_rolling.add_trace(go.Scatter(
    x=monthly2["month_dt"], y=monthly2["mean"],
    mode="lines+markers", name="Monatl. Ø OEE",
    line=dict(color="#DD8452", width=2.0),
    marker=dict(size=6, color="white", line=dict(color="#DD8452", width=1.8)),
    hovertemplate=(
        "<b>%{x|%B %Y}</b><br>Ø OEE: %{y:.1f} %<extra></extra>"
    ),
))

# Trace 4: rollierendes BDP → Treppenlinie (grün, durchgezogen, etwas breiter)
# Die Treppenlinie zeigt, wie das erreichbare Potenzial je Quartal variiert.
# Eine steigende Stufe = Maschine hat in diesem Quartal besser abgeschnitten.
# Eine sinkende Stufe = das Potenzial-Niveau war in diesem Quartal geringer
#   (z. B. wegen Umbau, Produktmix-Wechsel, saisonaler Belastung).
fig_rolling.add_trace(go.Scatter(
    x=step_x, y=step_y,
    mode="lines", name="Rollierendes BDP (90. Perz. je Quartal)",
    line=dict(color="#2CA02C", width=2.8),
    hoverinfo="skip",
))

# Invisible hover-Punkte je Quartal mit aussagekräftigem Tooltip
fig_rolling.add_trace(go.Scatter(
    x=bdp_quarterly["q_start"], y=bdp_quarterly["bdp_q90"],
    mode="markers",
    name="BDP je Quartal (Hover)",
    marker=dict(color="#2CA02C", size=10, opacity=0),   # unsichtbar
    showlegend=False,
    hovertemplate=(
        "<b>%{customdata}</b>"
        "<extra>Rollierendes BDP</extra>"
    ),
    customdata=bdp_quarterly["label"],
))

# Statische Gesamt-BDP-Referenzlinie (grün, gestrichelt) zum Vergleich
fig_rolling.add_hline(
    y=target_oee2,
    line=dict(color="#2CA02C", width=1.5, dash="dash"),
    annotation_text=f"Stat. BDP gesamt: {target_oee2:.1f} %",
    annotation_position="bottom right",
    annotation_font=dict(color="#2CA02C", size=11),
)

# Gesamtmittelwert (orange, gestrichelt)
fig_rolling.add_hline(
    y=overall_mean2,
    line=dict(color="#DD8452", width=1.5, dash="dash"),
    annotation_text=f"Ø {LINE2_LABEL}: {overall_mean2:.1f} %",
    annotation_position="bottom left",
    annotation_font=dict(color="#DD8452", size=11),
)

fig_rolling.update_layout(
    title=dict(
        text=(
            f"Rollierendes BDP-Potenzial — {LINE2_LABEL}<br>"
            f"<sup>{n_zero2} Leerläufe und {n_invalid2} Chargen ohne Output entfernt "
            f"| {len(df2)} Chargen in der Auswertung | Ziel-OEE wird quartalsweise neu berechnet</sup>"
        ),
        font=dict(size=16),
    ),
    xaxis=dict(
        title="Monat", tickformat="%b\n%Y", dtick="M1",
        showgrid=True, gridcolor="#e0e0e0",
    ),
    yaxis=dict(
        title="OEE [%]", range=[0, 100],
        showgrid=True, gridcolor="#e0e0e0",
    ),
    legend=dict(orientation="h", yanchor="bottom", y=1.02, xanchor="right", x=1),
    plot_bgcolor="white", paper_bgcolor="#f8f8f8",
    width=1000, height=500, hovermode="x unified",
)

fig_rolling.show()
```

## OEE-Komponenten: Cobas8000-1 vs. Cobas8000-2

| Faktor | Spalte | Frage, die er beantwortet |
|---|---|---|
| **Verfügbarkeit** (A) | `availability` | Wie oft stand die Anlage tatsächlich zur Verfügung? |
| **Leistungsgrad** (P) | `performance` | Wie schnell hat sie produziert, gemessen am Soll? |
| **Qualität** (Q) | `quality` | Wie viel des Outputs war in Ordnung? |
| **OEE (Ist)** | `oee` | Kombinierter Wert aus der Datenbank |
| **OEE (A×P×Q)** | *berechnet* | Kreuzprodukt – sollte mit Ist-OEE übereinstimmen |

```python
import pandas as pd
import numpy as np
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import warnings
warnings.filterwarnings("ignore")

COBAS = {
    "Cobas8000-1": {
        "line_name":  "Cobas8000-",
        "line_oid":   "C5CEB0CA-FC9E-4979-A2DB-6A77B6F330A7",
        "color":      "#4C72B0",
        "color_fill": "rgba(76, 114, 176, 0.15)",
    },
    "Cobas8000-2": {
        "line_name":  "Cobas8000-",
        "line_oid":   "074FADAA-59C9-420A-BE0E-6367E00BE429",
        "color":      "#DD8452",
        "color_fill": "rgba(221, 132, 82, 0.15)",
    },
}

COMP = {
    "availability_pct": {"name": "Verfügbarkeit", "color": "#4C72B0", "dash": None,  "width": 2.2},
    "performance_pct":  {"name": "Leistungsgrad", "color": "#DD8452", "dash": None,  "width": 2.2},
    "quality_pct":      {"name": "Qualität",      "color": "#55A868", "dash": None,  "width": 2.2},
    "oee_pct":          {"name": "OEE (Ist)",     "color": "#C44E52", "dash": None,  "width": 3.0},
    "derived_oee_pct":  {"name": "OEE (A×P×Q)",  "color": "#9370DB", "dash": "dot", "width": 1.8},
}

OEE_COLS = ["oee", "availability", "performance", "quality"]


def prepare_line(meta: dict) -> tuple:
    """Bereitet die Rohdaten einer Linie für die Visualisierung auf.

    Schritte: Filtern → Typen sichern → Start-/Stoppdatetime zusammenführen
              → NaN entfernen → Quantity-Filter → Wertebereichsprüfung OEE
              → Prozentskalierung → Komponenten-Cap (alle auf 0–100 %)
              → Chargendauer berechnen → zeitgewichtete Monatsaggregation
              → A×P×Q ableiten.
    """

    mask = (
        (df["line_name"] == meta["line_name"]) &
        (df["line_oid"]  == meta["line_oid"])
    )
    dfl = df[mask].copy()

    # Datentypen absichern.
    for col in OEE_COLS:
        dfl[col] = pd.to_numeric(dfl[col], errors="coerce")
    dfl["total_quantity"] = pd.to_numeric(dfl["total_quantity"], errors="coerce")

    # Startdatum + Uhrzeit aus zwei separaten CSV-Spalten zusammenführen.
    # "start"       = Datum    (z. B. "2024-07-15")
    # "Unnamed: 14" = Uhrzeit  (z. B. "08:23:00")
    dfl["start"] = pd.to_datetime(
        dfl["start"].astype(str) + " " + dfl["Unnamed: 14"].astype(str),
        errors="coerce",
    )
    # Stoppzeitpunkt analog zusammenführen ("stop" + "Unnamed: 16").
    dfl["stop_dt"] = pd.to_datetime(
        dfl["stop"].astype(str) + " " + dfl["Unnamed: 16"].astype(str),
        errors="coerce",
    )

    # Zeilen ohne OEE-Wert, Startzeit oder Stoppzeit sind nicht nutzbar.
    dfl = dfl.dropna(subset=["oee", "start", "stop_dt"])

    # Chargen ohne jede Produktion ausschließen.
    n_zero = (dfl["total_quantity"] == 0).sum()
    dfl = dfl[dfl["total_quantity"] > 0].copy()
    if n_zero > 0:
        print(f"  [{meta['line_name']} / {meta['line_oid'][:8]}] "
              f"{n_zero} Charge(n) entfernt: keine Produktionsmenge vorhanden (Leerläufe).")

    # Ungültige OEE-Werte entfernen.
    n_invalid = ((dfl["oee"] <= 0) | (dfl["oee"] > 1)).sum()
    dfl = dfl[(dfl["oee"] > 0) & (dfl["oee"] <= 1)].copy()
    if n_invalid > 0:
        print(f"  [{meta['line_name']} / {meta['line_oid'][:8]}] "
              f"{n_invalid} Charge(n) entfernt: kein Output erzielt oder ungültiger Datenwert.")

    # Alle Kennzahlen auf Prozent skalieren.
    for col in OEE_COLS:
        dfl[f"{col}_pct"] = dfl[col] * 100

    # ── Komponenten-Cap: alle drei Faktoren auf [0, 100 %] begrenzen ──────
    #
    # Hintergrund: Im Quellsystem existieren Einträge, bei denen availability,
    # performance oder quality physikalisch unmögliche Werte > 1 tragen
    # (z. B. quality bis 266, performance bis 614 — Stammdatenfehler).
    # Beim einfachen Mittelwert wurden diese Extremwerte durch viele Chargen
    # verdünnt. Bei der Zeitgewichtung kann eine einzige lange Ausreißer-Charge
    # einen ganzen Monat dominieren und zu sichtbaren Spitzen im Diagramm führen.
    # → Alle drei Komponenten werden einheitlich auf 100 % gekappt, nicht entfernt,
    #   weil der OEE-Wert der Charge weiterhin gültig ist.

    # Leistungsgrad: am häufigsten betroffen (Sollgeschwindigkeit zu niedrig hinterlegt)
    n_capped_perf = (dfl["performance_pct"] > 100).sum()
    if n_capped_perf > 0:
        print(f"  [{meta['line_name']} / {meta['line_oid'][:8]}] "
              f"{n_capped_perf} Charge(n): Leistungsgrad auf 100 % begrenzt (Stammdaten zu niedrig hinterlegt)")
    dfl["performance_pct"] = dfl["performance_pct"].clip(lower=0, upper=100)

    # Qualität: physikalisch max. 100 % — Werte darüber sind Berechnungsfehler
    n_capped_qual = (dfl["quality_pct"] > 100).sum()
    n_neg_qual    = (dfl["quality_pct"] < 0).sum()
    if n_capped_qual > 0 or n_neg_qual > 0:
        print(f"  [{meta['line_name']} / {meta['line_oid'][:8]}] "
              f"{n_capped_qual} Charge(n): Qualität auf 100 % begrenzt | "
              f"{n_neg_qual} Charge(n): Qualität auf 0 % begrenzt (Stammdatenfehler)")
    dfl["quality_pct"] = dfl["quality_pct"].clip(lower=0, upper=100)

    # Verfügbarkeit: analog auf [0, 100 %] — Daten sind bereits sauber (max=1.0),
    # Clip als defensive Absicherung gegen zukünftige Datenfehler.
    dfl["availability_pct"] = dfl["availability_pct"].clip(lower=0, upper=100)

    # n_capped für Rückgabe: Summe aller gecappten Komponenten
    n_capped = n_capped_perf

    # Chargendauer berechnen — Grundlage für die Zeitgewichtung.
    dfl["duration_min"] = (dfl["stop_dt"] - dfl["start"]).dt.total_seconds() / 60
    n_no_dur = (dfl["duration_min"] <= 0).sum()
    dfl = dfl[dfl["duration_min"] > 0].copy()
    if n_no_dur > 0:
        print(f"  [{meta['line_name']} / {meta['line_oid'][:8]}] "
              f"{n_no_dur} Charge(n) entfernt: keine verwertbare Chargendauer (Start ≥ Stop).")

    # Zeitgewichtete monatliche Aggregation:
    #   Monatsmittelwerte werden proportional zur Chargendauer gewichtet,
    #   damit lange Chargen den Monatsdurchschnitt stärker beeinflussen als kurze.
    dfl["month"] = dfl["start"].dt.to_period("M")
    pct_cols = [f"{c}_pct" for c in OEE_COLS]
    monthly = (
        dfl.groupby("month")
        .apply(lambda x: pd.Series({
            col: np.average(x[col], weights=x["duration_min"])
            for col in pct_cols
        }))
        .reset_index()
    )
    monthly["month_dt"] = monthly["month"].dt.to_timestamp()

    # Abgeleitetes OEE aus A × P × Q.
    monthly["derived_oee_pct"] = (
        monthly["availability_pct"] / 100
        * monthly["performance_pct"] / 100
        * monthly["quality_pct"]    / 100
        * 100
    )

    return dfl, monthly, n_zero, n_invalid, n_capped


# Beide Cobas-Linien aufbereiten.
print("Datenaufbereitung:\n")
cobas_data = {}
for label, meta in COBAS.items():
    df_clean, _monthly, _n_zero, _n_invalid, _n_capped = prepare_line(meta)
    cobas_data[label] = {
        "df_clean":  df_clean,
        "monthly":   _monthly,
        "meta":      meta,
        "n_zero":    _n_zero,
        "n_invalid": _n_invalid,
        "n_capped":  _n_capped,
    }

print("\nDatensatz nach Bereinigung (total_quantity > 0 · OEE ∈ [0,1]) + Komponenten-Cap:\n")
for label, d in cobas_data.items():
    dfc = d["df_clean"]
    m   = d["monthly"]
    oee_w = np.average(dfc["oee_pct"], weights=dfc["duration_min"])
    print(f"  {label}")
    print(f"    Chargen:               {len(dfc)}")
    print(f"    Zeitraum:              {m['month'].min()} – {m['month'].max()}  ({len(m)} Monate)")
    print(f"    Ø OEE (zeitgewichtet): {oee_w:.1f} %")
    print(f"    Ø Verfügb.:            {np.average(dfc['availability_pct'], weights=dfc['duration_min']):.1f} %")
    print(f"    Ø Leistung:            {np.average(dfc['performance_pct'],  weights=dfc['duration_min']):.1f} %  (nach Cap)")
    print(f"    Ø Qualität:            {np.average(dfc['quality_pct'],      weights=dfc['duration_min']):.1f} %  (nach Cap)")
    print()
```

```python
# Erstellt den OEE-Komponentengraph für eine einzelne Linie.
def plot_oee_components(label: str) -> go.Figure:

    d       = cobas_data[label]
    dfc     = d["df_clean"]
    monthly = d["monthly"]
    meta    = d["meta"]

    # Zeitgewichteter Gesamt-OEE als Referenzlinie.
    oee_avg = np.average(dfc["oee_pct"], weights=dfc["duration_min"])

    fig = go.Figure()

    # Einzelchargen als sehr kleine, fast transparente Punkte im Hintergrund.
    fig.add_trace(go.Scatter(
        x             = dfc["start"],
        y             = dfc["oee_pct"],
        mode          = "markers",
        name          = "Einzelchargen (OEE)",
        marker        = dict(color=meta["color"], size=3, opacity=0.12),
        hovertemplate = "Datum: %{x|%d.%m.%Y}<br>OEE: %{y:.1f} %<extra>Einzelcharge</extra>",
    ))

    # Zeitgewichtete monatliche Mittelwertlinie je OEE-Komponente.
    for comp_col, cfg in COMP.items():
        if comp_col not in monthly.columns:
            continue

        line_cfg = dict(color=cfg["color"], width=cfg["width"])
        if cfg["dash"]:
            line_cfg["dash"] = cfg["dash"]

        fig.add_trace(go.Scatter(
            x             = monthly["month_dt"],
            y             = monthly[comp_col],
            mode          = "lines+markers",
            name          = cfg["name"],
            line          = line_cfg,
            marker        = dict(size=7, color="white",
                                 line=dict(color=cfg["color"], width=2)),
            hovertemplate = (
                f"<b>%{{x|%B %Y}}</b><br>"
                f"{cfg['name']}: %{{y:.1f}} %"
                "<extra></extra>"
            ),
        ))

    # Horizontale Referenzlinie für den zeitgewichteten Gesamt-OEE.
    fig.add_hline(
        y                   = oee_avg,
        line                = dict(color="#C44E52", width=1.5, dash="dash"),
        annotation_text     = f"Ø OEE gesamt: {oee_avg:.1f} %",
        annotation_position = "top right",
        annotation_font     = dict(color="#C44E52", size=12),
    )

    n_months  = len(monthly)
    n_batches = len(dfc)

    fig.update_layout(
        title = dict(
            text = (
                f"OEE-Komponenten im Zeitverlauf — {label}<br>"
                f"<sup>{cobas_data[label]['n_zero']} Leerläufe und "
                f"{cobas_data[label]['n_invalid']} Chargen ohne verwertbaren OEE entfernt | "
                f"{cobas_data[label]['n_capped']} Leistungsgrad auf 100 % begrenzt "
                f"| {n_batches} Chargen in der Auswertung, {n_months} Monate "
                f"| Mittelwert zeitgewichtet nach Chargendauer</sup>"
            ),
            font = dict(size=16),
        ),
        xaxis = dict(
            title      = "Monat",
            tickformat = "%b\n%Y",
            dtick      = "M1",
            showgrid   = True,
            gridcolor  = "#e8e8e8",
        ),
        yaxis = dict(
            title     = "[%]",
            range     = [0, 100],
            tickvals  = list(range(0, 101, 10)),
            showgrid  = True,
            gridcolor = "#e8e8e8",
        ),
        legend = dict(
            orientation = "h",
            yanchor     = "bottom", y = 1.02,
            xanchor     = "right",  x = 1,
            font        = dict(size=12),
        ),
        plot_bgcolor  = "white",
        paper_bgcolor = "#f8f8f8",
        width         = 1100,
        height        = 500,
        hovermode     = "x unified",
    )

    return fig


# Graph für Cobas8000-1 rendern.
plot_oee_components("Cobas8000-1").show()
```

| **4 · Komponenten-Cap** | Chargen, bei denen Leistungsgrad oder Qualität über 100 % lagen — Wert wird auf 100 % begrenzt, Charge bleibt erhalten | Physikalisch nicht möglich — Stammdatenfehler im Quellsystem (z. B. zu niedrig hinterlegte Sollgeschwindigkeit) |

#### Was sagt das gepunktete OEE (A×P×Q)?

Die lila gepunktete Linie ist das aus den drei Faktoren manuell berechnete OEE (A×P×Q).
Sie liegt bei beiden Geräten konsistent über dem roten Ist-OEE. Das bedeutet: Das
Quellsystem berechnet das OEE nach einer konservativeren Methode — möglicherweise
fließen dort auch Zeiten ein, die in den Einzelkomponenten nicht vollständig erfasst
sind (z. B. nicht zugeordnete Stillstände). Für die relative Bewertung der Faktoren
untereinander ist das unerheblich; für eine absolute OEE-Aussage sollte die
Berechnungslogik des Quellsystems dokumentiert und geprüft werden.


```python
import pandas as pd
import warnings
warnings.filterwarnings("ignore")

# ── 7 · Materialanalyse — OEE-Schwankungen nach Materialnummer ───────────────
#
# Fragen:
#   1. Welche Materialien werden am häufigsten produziert? (Volumen)
#   2. Wie stark schwankt der OEE je Material? (σ)
#   3. Gibt es Materialien mit systematisch niedrigem OEE?
#
# Datenbasis: bereinigter Datensatz beider Cobas8000-Linien (aus cobas_data),
# Mittelwerte werden zeitgewichtet nach Chargendauer berechnet.

MATERIAL_COL = "material_number"
TOP_N        = 10

# Beide Linien zusammenführen, Linienzuordnung als Spalte erhalten
df_both = pd.concat(
    [d["df_clean"].assign(linie=label) for label, d in cobas_data.items()],
    ignore_index=True
)

# Chargen ohne Materialnummer ausschließen
n_missing = df_both[MATERIAL_COL].isna().sum()
if n_missing > 0:
    print(f"  {n_missing} Charge(n) ohne Materialnummer ausgeschlossen.\n")
df_both = df_both.dropna(subset=[MATERIAL_COL])

# Materialnummer als sauberer String (führende Nullen erhalten)
df_both[MATERIAL_COL] = df_both[MATERIAL_COL].astype(str).str.strip()

# Anfahrkontrollen ausschließen
EXCLUDE_MATERIALS = ["Anfahrkontrolle"]
n_excluded = df_both[MATERIAL_COL].isin(EXCLUDE_MATERIALS).sum()
if n_excluded > 0:
    print(f"  {n_excluded} Charge(n) als Anfahrkontrolle ausgeschlossen.\n")
df_both = df_both[~df_both[MATERIAL_COL].isin(EXCLUDE_MATERIALS)]

# ── Zeitgewichtete Aggregation je Materialnummer ───────────────────────────
# np.average(..., weights=duration_min) gewichtet jeden OEE-Wert proportional
# zur Chargendauer — längere Chargen fließen stärker in den Mittelwert ein.
mat_stats = (
    df_both.groupby(MATERIAL_COL)
    .apply(lambda x: pd.Series({
        "chargen":    len(x),
        "oee_mean":   np.average(x["oee_pct"],          weights=x["duration_min"]),
        "oee_std":    x["oee_pct"].std(),
        "avail_mean": np.average(x["availability_pct"], weights=x["duration_min"]),
        "perf_mean":  np.average(x["performance_pct"],  weights=x["duration_min"]),
        "qual_mean":  np.average(x["quality_pct"],      weights=x["duration_min"]),
        "qty_total":  x["total_quantity"].sum(),
    }))
    .reset_index()
)

# Einzelchargen-Materialien: σ = NaN → 0
mat_stats["oee_std"] = mat_stats["oee_std"].fillna(0)
# groupby().apply() speichert alle Spalten als float64 — chargen zurück auf int,
# damit Balken-Beschriftungen "19" statt "19.0" zeigen.
mat_stats["chargen"] = mat_stats["chargen"].astype(int)

# Absteigend nach Chargenzahl
mat_stats = mat_stats.sort_values("chargen", ascending=False).reset_index(drop=True)
mat_top   = mat_stats.head(TOP_N).copy()

# ── Ausgabe ────────────────────────────────────────────────────────────────
overall_both = np.average(df_both["oee_pct"], weights=df_both["duration_min"])

print(f"Materialanalyse — Cobas8000-1 & Cobas8000-2 kombiniert")
print(f"Gesamt-Ø OEE (zeitgewichtet) über alle Chargen:  {overall_both:.2f} %")
print(f"Eindeutige Materialien (bereinigt): {len(mat_stats)}")
print(f"Darstellung: Top {TOP_N} nach Chargenzahl (absteigend)\n")

disp = mat_top[[
    MATERIAL_COL, "chargen", "qty_total",
    "oee_mean", "oee_std",
    "avail_mean", "perf_mean", "qual_mean",
]].copy()

disp.columns = [
    "Material", "Chargen", "Gesamtmenge",
    "Ø OEE", "σ OEE",
    "Ø Verfügb.", "Ø Leistung", "Ø Qualität",
]
for col in disp.columns[3:]:
    disp[col] = disp[col].round(1)

print(disp.to_string(index=False))

# ── Auffälligkeiten ────────────────────────────────────────────────────────
print("\n── Auffälligkeiten (innerhalb Top 10) ───────────────────────────────")

low_thresh = mat_top["oee_mean"].quantile(0.25)
high_std   = mat_top["oee_std"].quantile(0.75)

low_oee = mat_top[mat_top["oee_mean"] <= low_thresh]
if not low_oee.empty:
    print(f"\n  Niedriger Ø OEE (≤ {low_thresh:.1f} %, unteres Quartil):")
    for _, r in low_oee.iterrows():
        print(f"    {r[MATERIAL_COL]:<15}  "
              f"Chargen: {int(r['chargen']):>4}  "
              f"Ø OEE: {r['oee_mean']:.1f} %  "
              f"σ: {r['oee_std']:.1f} %")

vol_high_std = mat_top[mat_top["oee_std"] >= high_std]
if not vol_high_std.empty:
    print(f"\n  Hohe OEE-Schwankung (σ ≥ {high_std:.1f} %, oberes Quartil):")
    for _, r in vol_high_std.iterrows():
        print(f"    {r[MATERIAL_COL]:<15}  "
              f"Chargen: {int(r['chargen']):>4}  "

              f"Ø OEE: {r['oee_mean']:.1f} %  "
              f"σ: {r['oee_std']:.1f} %")
```
