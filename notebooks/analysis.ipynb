```python
import pandas as pd
from pathlib import Path

# Die CSV-Datei liegt im Ordner "data" relativ zum Notebook.
# Falls das Notebook aus einem Unterordner gestartet wird, wird eine Ebene
# nach oben navigiert, damit der Pfad trotzdem stimmt.
csv_path = Path.cwd() / "data" / "dbo.vw_batch_kpis_converted.csv"
if not csv_path.exists():
    csv_path = Path.cwd().parent / "data" / "dbo.vw_batch_kpis_converted.csv"

# Die Datei ist semicolon-getrennt und verwendet das deutsche Dezimalkomma.
# utf-8-sig entfernt automatisch die BOM (Byte Order Mark), die Excel beim
# Export manchmal an den Dateianfang schreibt.
df = pd.read_csv(
    csv_path,
    sep=";",
    decimal=",",
    encoding="utf-8-sig",
    low_memory=False,
)

print("Shape:", df.shape)
df.head(20)
```

```python
import numpy as np
import warnings
warnings.filterwarnings("ignore")

# Beide Cobas8000-Geräte teilen sich denselben line_name in der Datenbank.
# Um gezielt nur ein Gerät auszuwählen, filtert man zusätzlich über die
# eindeutige line_oid. LINE_LABEL wird nur für Diagrammtitel verwendet.
LINE       = "Cobas8000-"
LINE_OID   = "C5CEB0CA-FC9E-4979-A2DB-6A77B6F330A7"  # entspricht Cobas8000-1
LINE_LABEL = "Cobas8000-1"

# Nur Datensätze der gewählten Linie und des gewählten Geräts behalten.
# .copy() verhindert den pandas SettingWithCopyWarning bei späteren Zuweisungen.
df_line = df[(df["line_name"] == LINE) & (df["line_oid"] == LINE_OID)].copy()

# OEE, Startdatum und Mengenspalte in ihre korrekten Datentypen umwandeln.
# total_quantity liegt in der CSV als String vor und muss explizit konvertiert werden.
# errors="coerce" wandelt nicht-parsbare Werte in NaN um statt einen Fehler zu werfen.
df_line["oee"]            = pd.to_numeric(df_line["oee"],            errors="coerce")
df_line["start"]          = pd.to_datetime(df_line["start"],         errors="coerce")
df_line["total_quantity"] = pd.to_numeric(df_line["total_quantity"], errors="coerce")

# Zeilen ohne OEE-Wert oder Startdatum sind für die Analyse nicht nutzbar.
df_line = df_line.dropna(subset=["oee", "start"])

# Chargen ohne jede Produktion explizit ausschließen.
# Einträge mit total_quantity = 0 sind Leerläufe, Kurzstarts oder reine
# Stillstandserfassungen – sie haben keine OEE-Aussagekraft und würden
# den Mittelwert nach unten verzerren.
n_zero = (df_line["total_quantity"] == 0).sum()
df_line = df_line[df_line["total_quantity"] > 0].copy()
if n_zero > 0:
    print(f"  {n_zero} Charge(n) mit total_quantity = 0 entfernt.")

# Physikalisch unmögliche OEE-Werte entfernen.
# Ein OEE-Wert außerhalb des Intervalls [0, 1] ist ein Berechnungsfehler
# im Quellsystem (z. B. fehlerhafte Stammdaten oder Systemfehler).
# OEE = 0 bleibt bewusst erhalten – eine Charge ohne Ausbringung ist ein
# valides, aussagekräftiges Ereignis und gehört zur Analyse.
# Ein pauschaler Quantil-Trim wird hier bewusst nicht angewendet:
# Die Daten liegen qualitätsgeprüft vor. Ein Trim würde echte
# Extremereignisse (Störungen, Stillstände) entfernen, die für die
# OEE-Bewertung relevant sind.
n_invalid = ((df_line["oee"] < 0) | (df_line["oee"] > 1)).sum()
df_line = df_line[(df_line["oee"] >= 0) & (df_line["oee"] <= 1)].copy()
if n_invalid > 0:
    print(f"  {n_invalid} Charge(n) mit OEE außerhalb [0, 1] entfernt (Berechnungsfehler).")

# OEE liegt im Datensatz als Dezimalzahl (0–1) vor, für die Darstellung
# rechnen wir auf Prozent um.
df_line["oee_pct"] = df_line["oee"] * 100

# Für den Zeitreihenverlauf aggregieren wir auf Monatsebene.
# dt.to_period("M") erzeugt eine Periode wie "2024-07", die wir zum Gruppieren nutzen.
df_line["month"] = df_line["start"].dt.to_period("M")
monthly = (
    df_line.groupby("month")["oee_pct"]
    .agg(mean="mean", std="std", count="count")
    .reset_index()
)

# to_timestamp() wandelt die Period zurück in einen echten Datetime-Wert,
# den matplotlib und plotly direkt als Zeitachse interpretieren können.
monthly["month_dt"] = monthly["month"].dt.to_timestamp()

# Monate mit nur einer Charge haben keine Standardabweichung (NaN) – wir setzen 0.
monthly["std"] = monthly["std"].fillna(0)

# Gesamtmittelwert direkt aus den Einzelchargen berechnen, nicht als Mittelwert
# der Monatsmittelwerte. Letzteres würde Monate mit wenigen Chargen übergewichten.
overall_mean = df_line["oee_pct"].mean()

print(f"Linie:             {LINE_LABEL}")
print(f"Datenpunkte:       {len(df_line)}")
print(f"Zeitraum:          {monthly['month'].min()} – {monthly['month'].max()}")
print(f"Anzahl Monate:     {len(monthly)}")
print(f"Gesamtmittelwert:  {overall_mean:.2f} %")
print()
print(monthly[["month", "count", "mean", "std"]].to_string(index=False))
```

# OEE-Werte Cobas8000-1: Bibliotheksvergleich

Alle drei Diagramme zeigen **dieselbe Darstellung**: monatlicher Mittelwert-OEE der Produktionslinie  
**Cobas8000-1**  + Gesamtmittelwert-Referenzlinie + ±1σ-Band.

| Eigenschaft | **Matplotlib** | **Seaborn** | **Plotly** |
|---|---|---|---|
| Stil-Ansatz | Manuell / Low-Level | Theme-basiert / deklarativ | Interaktiv / webbasiert |
| Fehlerband | `ax.fill_between` manuell | `errorbar="sd"` in einem Aufruf | `go.Scatter` fill-to-next |
| Einzelpunkte | `ax.scatter` | `stripplot` | `go.Box` / scatter overlay |
| Interaktivität | Keine (statisch) | Keine (statisch) | Hover, Zoom, Pan |
| Ausgabeformat | PNG/SVG | PNG/SVG | HTML (inline im Notebook) |

## 1 · Matplotlib — vollständige manuelle Kontrolle

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# Matplotlib arbeitet auf sehr niedriger Ebene: jedes Element des Diagramms
# (Fläche, Linie, Punkte, Beschriftungen) wird einzeln und explizit gezeichnet.
# Das gibt maximale Kontrolle, erfordert aber auch mehr Code.

# Hintergrund des gesamten Figure-Objekts
# Hintergrund der Zeichenfläche selbst

fig, ax = plt.subplots(figsize=(12, 5))
fig.patch.set_facecolor("#f8f8f8")   
ax.set_facecolor("#f0f0f0")          

# Hilfsvariablen für bessere Lesbarkeit

# Zeitachse (Monatsstempel)
# monatlicher OEE-Mittelwert
# monatliche Standardabweichung

x = monthly["month_dt"]   
y = monthly["mean"]       
s = monthly["std"]        

# Das ±1σ-Band visualisiert die Streuung innerhalb eines Monats.
# fill_between füllt den Bereich zwischen (y - s) und (y + s) mit einer
# halbtransparenten Fläche (alpha=0.18) in der Hauptfarbe.
ax.fill_between(x, y - s, y + s, color="#4C72B0", alpha=0.18, label="±1 Std.-Abweichung")

# Die Mittelwertlinie liegt optisch über dem Band (höherer zorder-Standard).
# Die hohlen Marker (weiß gefüllt) heben jeden Monatswert hervor.
ax.plot(x, y, color="#4C72B0", linewidth=2.2, marker="o", markersize=6,
        markerfacecolor="white", markeredgewidth=1.8, label="Monatl. Ø OEE")

# Die Einzelchargen werden als sehr kleine, fast transparente Punkte
# im Hintergrund (zorder=1) gezeichnet, um die Datengrundlage sichtbar zu machen.
batch_x = df_line["start"]
batch_y = df_line["oee_pct"]
ax.scatter(batch_x, batch_y, color="#4C72B0", alpha=0.08, s=8, zorder=1, label="Einzelchargen")

# Die horizontale Referenzlinie zeigt den chargenseitig gewichteten Gesamtmittelwert.
ax.axhline(overall_mean, color="#C44E52", linewidth=1.8, linestyle="--",
           label=f"Gesamtmittelwert: {overall_mean:.1f} %")

# Jeden Monatspunkt mit dem gerundeten OEE-Wert beschriften.
# xytext=(0, 7) verschiebt die Beschriftung 7 Punkte nach oben.
for _, row in monthly.iterrows():
    ax.annotate(f"{row['mean']:.0f}", xy=(row["month_dt"], row["mean"]),
                xytext=(0, 7), textcoords="offset points",
                ha="center", fontsize=7, color="#333333")

# X-Achse: jeden Monat als Tick, Format "Jan\n2025"
ax.xaxis.set_major_locator(mdates.MonthLocator())
ax.xaxis.set_major_formatter(mdates.DateFormatter("%b\n%Y"))
ax.set_xlabel("Monat", fontsize=11)
ax.set_ylabel("OEE [%]", fontsize=11)
ax.set_title(f"Monatlicher OEE-Verlauf — {LINE_LABEL} (Matplotlib)",
             fontsize=13, fontweight="bold")
ax.set_ylim(0, 100)
ax.legend(fontsize=9, loc="lower left")
ax.grid(axis="y", linestyle="--", alpha=0.4)

# Obere und rechte Rahmenlinie entfernen – wirkt aufgeräumter.
for spine in ["top", "right"]:
    ax.spines[spine].set_visible(False)

plt.tight_layout()
plt.show()
```

## 2 · Seaborn — Theme-basiert, deklarativ

```python
import seaborn as sns

# Seaborn setzt ein globales Theme, das alle folgenden matplotlib-Plots
# beeinflusst. "whitegrid" gibt ein sauberes weißes Raster, "muted" dämpft
# die Farben etwas ab. Am Ende der Zelle setzen wir die Defaults zurück.
sns.set_theme(style="whitegrid", palette="muted", font_scale=1.05)

# Seaborn erwartet "long-format" Daten: eine Zeile pro Beobachtung.
# Wir konvertieren die Period-Spalte in einen lesbaren String, weil Seaborn
# kategoriale X-Achsen besser mit Strings als mit Period-Objekten umgehen kann.
df_plot = df_line.copy()
df_plot["month_str"] = df_plot["month"].dt.strftime("%Y-%m")

# Die Monate müssen explizit sortiert übergeben werden, sonst ordnet Seaborn
# die Kategorien alphabetisch – was hier zufällig stimmt, aber nicht garantiert ist.
month_order = sorted(df_plot["month_str"].unique())

fig, ax = plt.subplots(figsize=(12, 5))

# stripplot streut die Einzelchargen als kleine, halbtransparente Punkte
# leicht zufällig (jitter=True) um die Monatskategorie herum, damit
# übereinanderliegende Punkte sichtbar bleiben.
sns.stripplot(
    data=df_plot, x="month_str", y="oee_pct",
    order=month_order,
    color="#4C72B0", alpha=0.12, size=3, jitter=True, ax=ax, zorder=1,
)

# pointplot berechnet Mittelwert und Fehlerbalken intern.
# errorbar="sd" zeichnet ±1 Standardabweichung als Fehlerbalken mit Caps.
# Das ersetzt in Seaborn die manuelle fill_between-Logik aus Matplotlib.
sns.pointplot(
    data=df_plot, x="month_str", y="oee_pct",
    order=month_order,
    errorbar="sd",
    color="#4C72B0",
    markers="o",
    linestyles="-",
    err_kws={"linewidth": 1.5, "alpha": 0.6},
    capsize=0.12,
    ax=ax,
    zorder=3,
    label="Monatl. Ø OEE ± 1σ",
)

# Referenzlinie für den Gesamtdurchschnitt – identisch zu Matplotlib,
# da Seaborn intern ebenfalls matplotlib-Axes verwendet.
ax.axhline(overall_mean, color="#c44e52", linewidth=1.8, linestyle="--",
           label=f"Gesamtmittelwert: {overall_mean:.1f} %", zorder=4)

# Da die X-Achse kategorial ist, müssen die Labels manuell gesetzt werden.
ax.set_xticklabels(month_order, rotation=45, ha="right", fontsize=8)
ax.set_xlabel("Monat")
ax.set_ylabel("OEE [%]")
ax.set_ylim(0, 100)
ax.set_title(f"Monatlicher OEE-Verlauf — {LINE_LABEL} (Seaborn)\n"
             f"Bereinigung: OEE ∈ [0,1] | {len(df_line)} Chargen",
             fontsize=13, fontweight="bold")
ax.legend(fontsize=9, loc="lower left")

plt.tight_layout()
plt.show()

# Theme zurücksetzen, damit nachfolgende Matplotlib-Zellen nicht beeinflusst werden.
sns.reset_defaults()
```

## 3 · Plotly — interaktiv, webbasiert

```python
import plotly.graph_objects as go

# Plotly erzeugt interaktive HTML-Diagramme, die direkt im Notebook rendern.
# Jedes visuelle Element (Scatter, Band, Linie) ist ein eigener "Trace",
# der dem Figure-Objekt hinzugefügt wird. Reihenfolge = Zeichenreihenfolge.

# Hilfsvariablen für die monatlichen Werte
x      = monthly["month_dt"]
y_mean = monthly["mean"]
y_std  = monthly["std"]
y_up   = y_mean + y_std   # obere Bandgrenze
y_lo   = y_mean - y_std   # untere Bandgrenze

fig = go.Figure()

# Trace 1: Einzelchargen im Hintergrund.
# opacity=0.15 macht sie kaum sichtbar, aber beim Hovern über einen Punkt
# erscheinen Datum und exakter OEE-Wert als Tooltip.
fig.add_trace(go.Scatter(
    x=df_line["start"],
    y=df_line["oee_pct"],
    mode="markers",
    name="Einzelchargen",
    marker=dict(color="#4C72B0", size=4, opacity=0.15),
    hovertemplate="Datum: %{x|%Y-%m-%d}<br>OEE: %{y:.1f} %<extra>Einzelcharge</extra>",
))

# Trace 2: ±1σ-Band als geschlossene Fläche.
# Plotly kennt kein fill_between, daher wird der Pfad manuell konstruiert:
# erst von links nach rechts entlang der oberen Grenze, dann rückwärts
# entlang der unteren Grenze – das ergibt eine geschlossene Kurve.
# fill="toself" füllt diese Fläche. Die Linie selbst ist unsichtbar (alpha=0).
fig.add_trace(go.Scatter(
    x=list(x) + list(x[::-1]),
    y=list(y_up) + list(y_lo[::-1]),
    fill="toself",
    fillcolor="rgba(76, 114, 176, 0.18)",
    line=dict(color="rgba(255,255,255,0)"),
    hoverinfo="skip",       # kein Tooltip für die Bandfläche selbst
    showlegend=True,
    name="±1 Std.-Abweichung",
))

# Trace 3: monatliche Mittelwertlinie.
# customdata bündelt Zusatzinfos (σ, Chargenzahl) pro Punkt, die im
# hovertemplate über %{customdata[i]} referenziert werden.
fig.add_trace(go.Scatter(
    x=x,
    y=y_mean,
    mode="lines+markers",
    name="Monatl. Ø OEE",
    line=dict(color="#4C72B0", width=2.5),
    marker=dict(size=8, color="white", line=dict(color="#4C72B0", width=2)),
    hovertemplate=(
        "<b>%{x|%B %Y}</b><br>"
        "Ø OEE: %{y:.1f} %<br>"
        "σ: %{customdata[0]:.1f} %<br>"
        "Chargen: %{customdata[1]}<extra></extra>"
    ),
    customdata=np.column_stack([y_std, monthly["count"]]),
))

# Horizontale Referenzlinie für den Gesamtmittelwert.
# add_hline ist eine Hilfsmethode, die intern ebenfalls einen Shape-Layer erzeugt.
fig.add_hline(
    y=overall_mean,
    line=dict(color="#C44E52", width=2, dash="dash"),
    annotation_text=f"Ø {overall_mean:.1f} %",
    annotation_position="top right",
    annotation_font=dict(color="#C44E52", size=12),
)

# Layout: Achsenbeschriftung, Gitter, Hintergrundfarben.
# hovermode="x unified" zeigt beim Hovern alle Traces an einem X-Wert
# gleichzeitig in einem gemeinsamen Tooltip.
fig.update_layout(
    title=dict(
        text=f"Monatlicher OEE-Verlauf — {LINE_LABEL} (Plotly)<br>"
             f"<sup>Bereinigung: OEE ∈ [0,1] | {len(df_line)} Chargen</sup>",
        font=dict(size=16),
    ),
    xaxis=dict(
        title="Monat",
        tickformat="%b\n%Y",
        dtick="M1",           # ein Tick pro Monat
        showgrid=True,
        gridcolor="#e0e0e0",
    ),
    yaxis=dict(
        title="OEE [%]",
        range=[0, 100],
        showgrid=True,
        gridcolor="#e0e0e0",
    ),
    legend=dict(orientation="h", yanchor="bottom", y=1.02, xanchor="right", x=1),
    plot_bgcolor="white",
    paper_bgcolor="#f8f8f8",
    width=1000,
    height=500,
    hovermode="x unified",
)

fig.show()
```

## 4 · Cobas8000-2 — Vergleich der Gegenlinie

```python
import plotly.graph_objects as go

# ── 5 · Cobas8000-2 — Potenzialermittlung (Best-Demonstrated-Practice) ───────

LINE2       = "Cobas8000-"
LINE2_OID   = "074FADAA-59C9-420A-BE0E-6367E00BE429"
LINE2_LABEL = "Cobas8000-2"

# Filtern, Typen konvertieren, NaN-Zeilen entfernen.
# total_quantity liegt in der CSV als String vor und muss explizit konvertiert werden.
df2 = df[(df["line_name"] == LINE2) & (df["line_oid"] == LINE2_OID)].copy()
df2["oee"]            = pd.to_numeric(df2["oee"],            errors="coerce")
df2["start"]          = pd.to_datetime(df2["start"],         errors="coerce")
df2["total_quantity"] = pd.to_numeric(df2["total_quantity"], errors="coerce")
df2 = df2.dropna(subset=["oee", "start"])

# Chargen ohne jede Produktion explizit ausschließen.
# Einträge mit total_quantity = 0 sind Leerläufe, Kurzstarts oder reine
# Stillstandserfassungen – sie haben keine OEE-Aussagekraft.
n_zero2 = (df2["total_quantity"] == 0).sum()
df2 = df2[df2["total_quantity"] > 0].copy()
if n_zero2 > 0:
    print(f"  {n_zero2} Charge(n) mit total_quantity = 0 entfernt.")

# Ungültige OEE-Werte entfernen.
# Ausgeschlossen werden:
#   oee <= 0 : Leerläufe ohne Ausbringung (OEE = 0) sowie negative Werte
#              (Berechnungsfehler). Diese würden die Visualisierung nach
#              unten verzerren und sind für die BDP-Analyse nicht relevant.
#   oee > 1  : Physikalisch unmöglich – Berechnungsfehler im Quellsystem.
n_invalid2 = ((df2["oee"] <= 0) | (df2["oee"] > 1)).sum()
df2 = df2[(df2["oee"] > 0) & (df2["oee"] <= 1)].copy()
if n_invalid2 > 0:
    print(f"  {n_invalid2} Charge(n) mit OEE ≤ 0 oder > 1 entfernt.")

# clip(upper=100) verhindert, dass Floating-Point-Ungenauigkeiten
# (z. B. oee = 1.0000001) zu oee_pct minimal über 100 führen.
df2["oee_pct"] = (df2["oee"] * 100).clip(upper=100)

# Monatliche Aggregation
df2["month"] = df2["start"].dt.to_period("M")
monthly2 = (
    df2.groupby("month")["oee_pct"]
    .agg(mean="mean", std="std", count="count")
    .reset_index()
)
monthly2["month_dt"] = monthly2["month"].dt.to_timestamp()
monthly2["std"] = monthly2["std"].fillna(0)

# Chargenseitig gewichteter Gesamtmittelwert (Status Quo)
overall_mean2 = df2["oee_pct"].mean()

# ── Potenzialberechnung (Best-Demonstrated-Practice) ─────────────────────────
# Zielwert: Das 90. Perzentil des bereinigten Datensatzes.
# BDP = Best Demonstrated Practice: der beste nachweislich erreichte Zustand
# der eigenen Anlage – 10 % der Chargen haben diesen Wert bereits erzielt.
# Hinweis: Dieser Perzentil-Wert dient als Zielgröße – er ist kein Filter.
target_oee2   = df2["oee_pct"].quantile(0.90)

# Identifiziertes ungenutztes Potenzial (Differenz aus Zielwert und Ist-Zustand)
oee_potential = target_oee2 - overall_mean2

# ── Golden Batches ────────────────────────────────────────────────────────────
# Eine „Golden Batch" ist jede Charge, die den BDP-Zielwert (90. Perzentil)
# erreicht oder überschritten hat. Diese Chargen beweisen, dass der Zielwert
# unter realen Bedingungen erreichbar ist – sie sind die Lernvorbilder für das
# Shopfloor-Meeting. In der Visualisierung werden sie als grüne Punkte
# minimal stärker hervorgehoben als die Standard-Chargen.
mask_golden  = df2["oee_pct"] >= target_oee2
df2_standard = df2[~mask_golden]
df2_golden   = df2[mask_golden]
n_golden     = int(mask_golden.sum())
# ─────────────────────────────────────────────────────────────────────────────

print(f"Linie:            {LINE2_LABEL}")
print(f"Datenpunkte:      {len(df2)}")
print(f"Zeitraum:         {monthly2['month'].min()} – {monthly2['month'].max()}")
print(f"Gesamtmittelwert: {overall_mean2:.2f} %")
print(f"BDP-Zielwert (90. Perz.): {target_oee2:.2f} %")
print(f"Identifiziertes Potenzial: +{oee_potential:.2f} Prozentpunkte")
print(f"Golden Batches (≥ BDP-Ziel): {n_golden} von {len(df2)} Chargen "
      f"({n_golden / len(df2) * 100:.1f} %)")

# Hilfsvariablen für den Plot
x2     = monthly2["month_dt"]
y2     = monthly2["mean"]
y2_std = monthly2["std"]
y2_up  = y2 + y2_std
y2_lo  = y2 - y2_std

fig2 = go.Figure()

# Trace 1a: Standard-Chargen (orange, blass) – alle Chargen unterhalb des BDP-Ziels.
fig2.add_trace(go.Scatter(
    x=df2_standard["start"], y=df2_standard["oee_pct"],
    mode="markers", name="Einzelchargen",
    marker=dict(color="#DD8452", size=4, opacity=0.22),
    hovertemplate="Datum: %{x|%Y-%m-%d}<br>OEE: %{y:.1f} %<extra>Einzelcharge</extra>",
))

# Trace 1b: Golden Batches (grün, Kreis) – gleiche Größe (size 4), aber opaker
# (0.45 vs. 0.22) und grün statt orange. Subtile Hervorhebung ohne das
# Gesamtbild zu dominieren.
fig2.add_trace(go.Scatter(
    x=df2_golden["start"], y=df2_golden["oee_pct"],
    mode="markers", name=f"Golden Batch (≥ {target_oee2:.0f} %)",
    marker=dict(
        color="#2CA02C", size=4, opacity=0.45,
        symbol="circle",
    ),
    hovertemplate=(
        "<b>Golden Batch</b><br>"
        "Datum: %{x|%Y-%m-%d}<br>"
        "OEE: %{y:.1f} %"
        "<extra>Golden Batch</extra>"
    ),
))

# Trace 2: ±1σ-Band (geschlossene Fläche)
fig2.add_trace(go.Scatter(
    x=list(x2) + list(x2[::-1]),
    y=list(y2_up) + list(y2_lo[::-1]),
    fill="toself",
    fillcolor="rgba(221, 132, 82, 0.18)",
    line=dict(color="rgba(255,255,255,0)"),
    hoverinfo="skip", showlegend=True,
    name="±1 Std.-Abweichung",
))

# Trace 3: monatliche Mittelwertlinie mit Hover-Tooltip
fig2.add_trace(go.Scatter(
    x=x2, y=y2,
    mode="lines+markers", name="Monatl. Ø OEE",
    line=dict(color="#DD8452", width=2.5),
    marker=dict(size=8, color="white", line=dict(color="#DD8452", width=2)),
    hovertemplate=(
        "<b>%{x|%B %Y}</b><br>"
        "Ø OEE: %{y:.1f} %<br>"
        "σ: %{customdata[0]:.1f} %<br>"
        "Chargen: %{customdata[1]}<extra></extra>"
    ),
    customdata=np.column_stack([y2_std, monthly2["count"]]),
))

# Referenzlinie: BDP-Zielwert / 90. Perzentil (grün)
fig2.add_hline(
    y=target_oee2,
    line=dict(color="#2CA02C", width=2.5),
    annotation_text=f"BDP-Ziel (90. Perz.): {target_oee2:.1f} %",
    annotation_position="top left",
    annotation_font=dict(color="#2CA02C", size=12),
)

# Referenzlinie: eigener Gesamtmittelwert / Status Quo (orange, gestrichelt)
fig2.add_hline(
    y=overall_mean2,
    line=dict(color="#DD8452", width=2, dash="dash"),
    annotation_text=f"Ø {LINE2_LABEL}: {overall_mean2:.1f} %",
    annotation_position="bottom left",
    annotation_font=dict(color="#DD8452", size=11),
)

fig2.update_layout(
    title=dict(
        text=(
            f"Monatlicher OEE-Verlauf & Potenzialermittlung — {LINE2_LABEL}<br>"
            f"<sup>Bereinigung: OEE ∈ (0,1] | BDP-Zielwert: 90. Perzentil | "
            f"Identifiziertes OEE-Potenzial: +{oee_potential:.1f} %-Punkte | "
            f"{n_golden} Golden Batches ({n_golden / len(df2) * 100:.1f} %)</sup>"
        ),
        font=dict(size=16),
    ),
    xaxis=dict(
        title="Monat", tickformat="%b\n%Y", dtick="M1",
        showgrid=True, gridcolor="#e0e0e0",
    ),
    yaxis=dict(
        title="OEE [%]", range=[0, 100],
        showgrid=True, gridcolor="#e0e0e0",
    ),
    legend=dict(orientation="h", yanchor="bottom", y=1.02, xanchor="right", x=1),
    plot_bgcolor="white", paper_bgcolor="#f8f8f8",
    width=1000, height=500, hovermode="x unified",
)

fig2.show()
```

```python
import plotly.graph_objects as go

# ── 6 · Cobas8000-2 — rollierendes BDP-Potenzial (dynamisches 90. Perzentil) ─
#
# Statt einer einzigen statischen Linie über 24 Monate wird das 90. Perzentil
# quartalsweise neu berechnet. Das ergibt eine Treppenlinie, die zeigt, wie
# sich das erreichbare Potenzial der Anlage im Zeitverlauf verändert hat.
#
# Methodik:
#   - Perioden: Q1 2024, Q2 2024, … Q4 2025, Q1 2026
#   - Je Quartal: 90. Perzentil aller Chargen dieses Quartals
#   - Darstellung als Treppenstufe (step="post"): der Wert gilt ab Quartalsbeginn
#     bis das nächste Quartal beginnt.
#
# Warum Quartale und nicht Monate?
#   Ein monatliches Perzentil ist bei oft nur 20–40 Chargen pro Monat statistisch
#   instabil (ein einzelner Ausreißer bewegt das 90. Perz. stark). Quartale
#   bündeln ~60–120 Chargen und liefern robustere Schätzwerte.

# df2 und target_oee2 / overall_mean2 stammen aus der vorherigen Zelle.
# Sicherheitshalber explizit referenzieren:
_df = df2.copy()
_df["quarter"] = _df["start"].dt.to_period("Q")

# 90. Perzentil je Quartal berechnen.
# min_count=1 stellt sicher, dass Quartale mit nur wenigen Chargen nicht
# stillschweigend übersprungen werden.
bdp_quarterly = (
    _df.groupby("quarter")["oee_pct"]
    .quantile(0.90)
    .reset_index()
    .rename(columns={"oee_pct": "bdp_q90"})
)

# Quartal → Zeitstempel (Quartalsbeginn) für die X-Achse
bdp_quarterly["q_start"] = bdp_quarterly["quarter"].dt.to_timestamp("D", how="start")

# Quartal-Ende für die Treppenstufe: letzter Tag des Quartals als nächster
# Quartalsbeginn. So reicht jede Stufe exakt bis zum nächsten Quartalswechsel.
bdp_quarterly["q_end"] = bdp_quarterly["quarter"].dt.to_timestamp("D", how="end")

# Für eine saubere Treppenlinie in Plotly: jeden Wert zweimal eintragen –
# einmal am Anfang der Periode, einmal am Ende. Das ergibt exakt waagerechte
# Stufen ohne diagonale Verbindungslinien.
step_x, step_y = [], []
for _, row in bdp_quarterly.iterrows():
    step_x += [row["q_start"], row["q_end"]]
    step_y += [row["bdp_q90"], row["bdp_q90"]]

# Annotationstext je Quartal für den Hover-Tooltip
bdp_quarterly["label"] = bdp_quarterly.apply(
    lambda r: f"Q{r['quarter'].quarter} {r['quarter'].year}: {r['bdp_q90']:.1f} %", axis=1
)

print("Rollierendes BDP-Potenzial (90. Perzentil je Quartal):\n")
for _, row in bdp_quarterly.iterrows():
    n = int(_df[_df["quarter"] == row["quarter"]]["oee_pct"].count())
    print(f"  {row['label']}  ({n} Chargen)")

print(f"\n  Statisches Gesamt-BDP zum Vergleich: {target_oee2:.1f} %")
print(f"  Gesamtmittelwert Status Quo:          {overall_mean2:.1f} %")

# ─── Diagramm ────────────────────────────────────────────────────────────────
fig_rolling = go.Figure()

# Trace 1: Einzelchargen (Standard, orange blass)
fig_rolling.add_trace(go.Scatter(
    x=df2["start"], y=df2["oee_pct"],
    mode="markers", name="Einzelchargen",
    marker=dict(color="#DD8452", size=4, opacity=0.22),
    hovertemplate="Datum: %{x|%Y-%m-%d}<br>OEE: %{y:.1f} %<extra>Einzelcharge</extra>",
))

# Trace 2: Golden Batches (grün, leicht opaker)
mask_golden_r = df2["oee_pct"] >= target_oee2
fig_rolling.add_trace(go.Scatter(
    x=df2[mask_golden_r]["start"], y=df2[mask_golden_r]["oee_pct"],
    mode="markers", name=f"Golden Batch (≥ stat. BDP {target_oee2:.0f} %)",
    marker=dict(color="#2CA02C", size=4, opacity=0.45, symbol="circle"),
    hovertemplate=(
        "<b>Golden Batch</b><br>Datum: %{x|%Y-%m-%d}<br>OEE: %{y:.1f} %"
        "<extra>Golden Batch</extra>"
    ),
))

# Trace 3: monatliche Mittelwertlinie
fig_rolling.add_trace(go.Scatter(
    x=monthly2["month_dt"], y=monthly2["mean"],
    mode="lines+markers", name="Monatl. Ø OEE",
    line=dict(color="#DD8452", width=2.0),
    marker=dict(size=6, color="white", line=dict(color="#DD8452", width=1.8)),
    hovertemplate=(
        "<b>%{x|%B %Y}</b><br>Ø OEE: %{y:.1f} %<extra></extra>"
    ),
))

# Trace 4: rollierendes BDP → Treppenlinie (grün, durchgezogen, etwas breiter)
# Die Treppenlinie zeigt, wie das erreichbare Potenzial je Quartal variiert.
# Eine steigende Stufe = Maschine hat in diesem Quartal besser abgeschnitten.
# Eine sinkende Stufe = das Potenzial-Niveau war in diesem Quartal geringer
#   (z. B. wegen Umbau, Produktmix-Wechsel, saisonaler Belastung).
fig_rolling.add_trace(go.Scatter(
    x=step_x, y=step_y,
    mode="lines", name="Rollierendes BDP (90. Perz. je Quartal)",
    line=dict(color="#2CA02C", width=2.8),
    hoverinfo="skip",
))

# Invisible hover-Punkte je Quartal mit aussagekräftigem Tooltip
fig_rolling.add_trace(go.Scatter(
    x=bdp_quarterly["q_start"], y=bdp_quarterly["bdp_q90"],
    mode="markers",
    name="BDP je Quartal (Hover)",
    marker=dict(color="#2CA02C", size=10, opacity=0),   # unsichtbar
    showlegend=False,
    hovertemplate=(
        "<b>%{customdata}</b>"
        "<extra>Rollierendes BDP</extra>"
    ),
    customdata=bdp_quarterly["label"],
))

# Statische Gesamt-BDP-Referenzlinie (grün, gestrichelt) zum Vergleich
fig_rolling.add_hline(
    y=target_oee2,
    line=dict(color="#2CA02C", width=1.5, dash="dash"),
    annotation_text=f"Stat. BDP gesamt: {target_oee2:.1f} %",
    annotation_position="bottom right",
    annotation_font=dict(color="#2CA02C", size=11),
)

# Gesamtmittelwert (orange, gestrichelt)
fig_rolling.add_hline(
    y=overall_mean2,
    line=dict(color="#DD8452", width=1.5, dash="dash"),
    annotation_text=f"Ø {LINE2_LABEL}: {overall_mean2:.1f} %",
    annotation_position="bottom left",
    annotation_font=dict(color="#DD8452", size=11),
)

fig_rolling.update_layout(
    title=dict(
        text=(
            f"Rollierendes BDP-Potenzial — {LINE2_LABEL}<br>"
            f"<sup>Quartalsweises 90. Perzentil  ·  {len(df2)} Chargen  ·  "
            f"Zeigt Veränderung des erreichbaren Potenzials im Zeitverlauf</sup>"
        ),
        font=dict(size=16),
    ),
    xaxis=dict(
        title="Monat", tickformat="%b\n%Y", dtick="M1",
        showgrid=True, gridcolor="#e0e0e0",
    ),
    yaxis=dict(
        title="OEE [%]", range=[0, 100],
        showgrid=True, gridcolor="#e0e0e0",
    ),
    legend=dict(orientation="h", yanchor="bottom", y=1.02, xanchor="right", x=1),
    plot_bgcolor="white", paper_bgcolor="#f8f8f8",
    width=1000, height=500, hovermode="x unified",
)

fig_rolling.show()
```

|---|---|---|

| **OEE (A×P×Q)** | *berechnet* | Kreuzprodukt – sollte mit Ist-OEE übereinstimmen |

```python
import pandas as pd
import numpy as np
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import warnings
warnings.filterwarnings("ignore")

# Beide Cobas8000-Geräte tragen denselben line_name in der Datenbank.
# Sie werden ausschließlich über ihre eindeutige line_oid unterschieden.
# color und color_fill werden später für die Plotly-Traces verwendet.
COBAS = {
    "Cobas8000-1": {
        "line_name":  "Cobas8000-",
        "line_oid":   "C5CEB0CA-FC9E-4979-A2DB-6A77B6F330A7",
        "color":      "#4C72B0",
        "color_fill": "rgba(76, 114, 176, 0.15)",
    },
    "Cobas8000-2": {
        "line_name":  "Cobas8000-",
        "line_oid":   "074FADAA-59C9-420A-BE0E-6367E00BE429",
        "color":      "#DD8452",
        "color_fill": "rgba(221, 132, 82, 0.15)",
    },
}

# Farb- und Stilkonfiguration je OEE-Komponente.
# Einmal hier definiert und in allen Diagrammen wiederverwendet,
# damit jede Kurve immer denselben Farbton trägt.
# dash=None bedeutet durchgezogene Linie; "dot" erzeugt eine gepunktete Linie.
COMP = {
    "availability_pct": {"name": "Verfügbarkeit", "color": "#4C72B0", "dash": None,  "width": 2.2},
    "performance_pct":  {"name": "Leistungsgrad", "color": "#DD8452", "dash": None,  "width": 2.2},
    "quality_pct":      {"name": "Qualität",      "color": "#55A868", "dash": None,  "width": 2.2},
    "oee_pct":          {"name": "OEE (Ist)",     "color": "#C44E52", "dash": None,  "width": 3.0},
    "derived_oee_pct":  {"name": "OEE (A×P×Q)",  "color": "#9370DB", "dash": "dot", "width": 1.8},
}

# Spalten, die aus dem Rohdatensatz gelesen und verarbeitet werden.
OEE_COLS = ["oee", "availability", "performance", "quality"]


# Bereitet die Rohdaten einer Linie für die Visualisierung auf.
# Schritte: Filtern → Typen sichern → NaN entfernen → Quantity-Filter
#           → Wertebereichsprüfung OEE ∈ [0,1] → Prozentskalierung
#           → Performance-Cap → Monatsaggregation → A×P×Q
def prepare_line(meta: dict) -> tuple:

    # Nur Zeilen der gewünschten Linie und des gewünschten Geräts behalten.
    mask = (
        (df["line_name"] == meta["line_name"]) &
        (df["line_oid"]  == meta["line_oid"])
    )
    dfl = df[mask].copy()

    # Datentypen absichern, da Rohdaten teils als gemischter Typ vorliegen.
    # total_quantity liegt in der CSV als String vor und muss explizit konvertiert werden.
    # errors="coerce" wandelt unlesbare Werte in NaN um statt einen Fehler zu werfen.
    for col in OEE_COLS:
        dfl[col] = pd.to_numeric(dfl[col], errors="coerce")
    dfl["start"]          = pd.to_datetime(dfl["start"],         errors="coerce")
    dfl["total_quantity"] = pd.to_numeric(dfl["total_quantity"], errors="coerce")

    # Zeilen ohne OEE-Wert oder Datum sind für Zeitreihen unbrauchbar.
    dfl = dfl.dropna(subset=["oee", "start"])

    # Chargen ohne jede Produktion explizit ausschließen.
    # Einträge mit total_quantity = 0 sind Leerläufe, Kurzstarts oder reine
    # Stillstandserfassungen – sie haben keine OEE-Aussagekraft und würden
    # den Mittelwert nach unten verzerren.
    n_zero = (dfl["total_quantity"] == 0).sum()
    dfl = dfl[dfl["total_quantity"] > 0].copy()
    if n_zero > 0:
        print(f"  [{meta['line_name']} / {meta['line_oid'][:8]}] "
              f"{n_zero} Charge(n) mit total_quantity = 0 entfernt.")

    # Physikalisch unmögliche OEE-Werte entfernen.
    # Ein OEE-Wert außerhalb des Intervalls [0, 1] ist ein Berechnungsfehler
    # im Quellsystem (z. B. fehlerhafte Stammdaten oder Systemfehler).
    # OEE = 0 bleibt bewusst erhalten – eine Charge ohne Ausbringung ist ein
    # valides, aussagekräftiges Ereignis und gehört zur Analyse.
    # Ein pauschaler Quantil-Trim wird hier bewusst nicht angewendet:
    # Die Daten liegen qualitätsgeprüft vor. Ein Trim würde echte
    # Extremereignisse (Störungen, Stillstände) entfernen, die für die
    # OEE-Analyse inhaltlich relevant sind und den Istzustand korrekt abbilden.
    n_invalid = ((dfl["oee"] < 0) | (dfl["oee"] > 1)).sum()
    dfl = dfl[(dfl["oee"] >= 0) & (dfl["oee"] <= 1)].copy()
    if n_invalid > 0:
        print(f"  [{meta['line_name']} / {meta['line_oid'][:8]}] "
              f"{n_invalid} Charge(n) mit OEE außerhalb [0, 1] entfernt "
              f"(Berechnungsfehler im Quellsystem).")

    # Alle Kennzahlen von Dezimalzahl (0–1) auf Prozent (0–100) skalieren.
    for col in OEE_COLS:
        dfl[f"{col}_pct"] = dfl[col] * 100

    # Leistungsgrad-Cap bei 100 %.
    # Werte über 100 % sind physikalisch nicht möglich und entstehen, wenn die
    # theoretische Ausbringungsrate im Stammdatensystem zu niedrig hinterlegt ist.
    # Betroffene Chargen werden auf 100 % begrenzt, nicht entfernt,
    # da der restliche OEE-Wert der Charge weiterhin gültig ist.
    n_capped = (dfl["performance_pct"] > 100).sum()
    if n_capped > 0:
        print(f"  [{meta['line_name']} / {meta['line_oid'][:8]}] "
              f"{n_capped} Charge(n) mit Leistungsgrad > 100 % auf 100 % begrenzt "
              f"(Stammdatenfehler: theoretische Rate zu niedrig hinterlegt)")
    dfl["performance_pct"] = dfl["performance_pct"].clip(upper=100)

    # Monatliche Mittelwerte aggregieren.
    # to_period("M") gruppiert nach Kalendermonat; to_timestamp() wandelt
    # die Period danach wieder in einen echten Datetime-Wert um, den Plotly
    # direkt als Zeitachse interpretieren kann.
    dfl["month"] = dfl["start"].dt.to_period("M")
    pct_cols = [f"{c}_pct" for c in OEE_COLS]
    monthly = dfl.groupby("month")[pct_cols].mean().reset_index()
    monthly["month_dt"] = monthly["month"].dt.to_timestamp()

    # Abgeleitetes OEE aus den drei Einzelfaktoren berechnen: A × P × Q.
    # Stimmt dieser Wert nicht mit dem Ist-OEE überein, liegt eine abweichende
    # Berechnungslogik im Quellsystem vor.
    monthly["derived_oee_pct"] = (
        monthly["availability_pct"] / 100
        * monthly["performance_pct"] / 100
        * monthly["quality_pct"]    / 100
        * 100
    )

    return dfl, monthly


# Beide Cobas-Linien aufbereiten und Ergebnisse in einem Dictionary speichern.
print("Datenaufbereitung:\n")
cobas_data = {}
for label, meta in COBAS.items():
    df_clean, _monthly = prepare_line(meta)
    cobas_data[label] = {
        "df_clean": df_clean,
        "monthly":  _monthly,
        "meta":     meta,
    }

# Datenbasis als Überblick ausgeben.
print("\nDatensatz nach Bereinigung (total_quantity > 0 · OEE ∈ [0,1]) + Performance-Cap:\n")
for label, d in cobas_data.items():
    dfc = d["df_clean"]
    m   = d["monthly"]
    print(f"  {label}")
    print(f"    Chargen:       {len(dfc)}")
    print(f"    Zeitraum:      {m['month'].min()} – {m['month'].max()}  ({len(m)} Monate)")
    print(f"    Ø OEE:         {dfc['oee_pct'].mean():.1f} %")
    print(f"    Ø Verfügb.:    {dfc['availability_pct'].mean():.1f} %")
    print(f"    Ø Leistung:    {dfc['performance_pct'].mean():.1f} %  (nach Cap)")
    print(f"    Ø Qualität:    {dfc['quality_pct'].mean():.1f} %")
    print()
```

```python
# Erstellt den OEE-Komponentengraph für eine einzelne Linie.
# Der Graph enthält: Einzelchargen-Scatter im Hintergrund, monatliche
# Mittelwertlinien je Komponente sowie eine horizontale OEE-Referenzlinie.
def plot_oee_components(label: str) -> go.Figure:

    # Vorberechnete Daten für die gewählte Linie laden.
    d       = cobas_data[label]
    dfc     = d["df_clean"]
    monthly = d["monthly"]
    meta    = d["meta"]
    oee_avg = dfc["oee_pct"].mean()

    fig = go.Figure()

    # Einzelchargen als sehr kleine, fast transparente Punkte im Hintergrund.
    # Sie geben ein Gefühl für die tatsächliche Streuung innerhalb der Monate,
    # ohne die Trendlinien im Vordergrund zu überlagern.
    # Beim Hovern über einen Punkt erscheinen Datum und genauer OEE-Wert.
    fig.add_trace(go.Scatter(
        x             = dfc["start"],
        y             = dfc["oee_pct"],
        mode          = "markers",
        name          = "Einzelchargen (OEE)",
        marker        = dict(color=meta["color"], size=3, opacity=0.12),
        hovertemplate = "Datum: %{x|%d.%m.%Y}<br>OEE: %{y:.1f} %<extra>Einzelcharge</extra>",
    ))

    # Monatliche Mittelwertlinie je OEE-Komponente.
    # Die Reihenfolge der Schleife bestimmt die Zeichenreihenfolge:
    # OEE-Ist ist das letzte Element in COMP und liegt dadurch optisch oben.
    for comp_col, cfg in COMP.items():
        if comp_col not in monthly.columns:
            continue

        # Linienstil aufbauen; dash wird nur gesetzt, wenn explizit angegeben.
        line_cfg = dict(color=cfg["color"], width=cfg["width"])
        if cfg["dash"]:
            line_cfg["dash"] = cfg["dash"]

        # Hohle Marker mit weißem Kern heben die Monatsstützpunkte hervor,
        # ohne die Linie optisch zu unterbrechen.
        fig.add_trace(go.Scatter(
            x             = monthly["month_dt"],
            y             = monthly[comp_col],
            mode          = "lines+markers",
            name          = cfg["name"],
            line          = line_cfg,
            marker        = dict(size=7, color="white",
                                 line=dict(color=cfg["color"], width=2)),
            hovertemplate = (
                f"<b>%{{x|%B %Y}}</b><br>"
                f"{cfg['name']}: %{{y:.1f}} %"
                "<extra></extra>"
            ),
        ))

    # Horizontale Referenzlinie für den chargenseitig gewichteten Gesamt-OEE.
    # So ist auf einen Blick erkennbar, ob ein Monat über oder unter dem
    # langjährigen Durchschnitt liegt.
    fig.add_hline(
        y                   = oee_avg,
        line                = dict(color="#C44E52", width=1.5, dash="dash"),
        annotation_text     = f"Ø OEE gesamt: {oee_avg:.1f} %",
        annotation_position = "top right",
        annotation_font     = dict(color="#C44E52", size=12),
    )

    n_months  = len(monthly)
    n_batches = len(dfc)

    # Layout-Einstellungen: Titel, Achsen, Legende, Hintergrundfarben.
    fig.update_layout(
        title = dict(
            text = (
                f"OEE-Komponenten im Zeitverlauf — {label}<br>"
                f"<sup>Monatliche Mittelwerte  ·  {n_batches} Chargen  ·  "
                f"{n_months} Monate  ·  Bereinigung: OEE ∈ [0,1]  ·  "
                f"Leistungsgrad auf max. 100 % begrenzt</sup>"
            ),
            font = dict(size=16),
        ),
        xaxis = dict(
            title      = "Monat",
            tickformat = "%b\n%Y",
            dtick      = "M1",
            showgrid   = True,
            gridcolor  = "#e8e8e8",
        ),
        yaxis = dict(
            title     = "[%]",
            range     = [0, 100],
            tickvals  = list(range(0, 101, 10)),
            showgrid  = True,
            gridcolor = "#e8e8e8",
        ),
        legend = dict(
            orientation = "h",
            yanchor     = "bottom", y = 1.02,
            xanchor     = "right",  x = 1,
            font        = dict(size=12),
        ),
        plot_bgcolor  = "white",
        paper_bgcolor = "#f8f8f8",
        width         = 1100,
        height        = 500,
        hovermode     = "x unified",
    )

    return fig


# Graph für Cobas8000-1 rendern.
plot_oee_components("Cobas8000-1").show()
```

```python
# Graph für Cobas8000-2 rendern.
# Nutzt dieselbe Funktion wie Cobas8000-1 – nur das Label ändert sich,
# der Rest wird automatisch aus cobas_data bezogen.
plot_oee_components("Cobas8000-2").show()
```

## Interpretation der Diagramme

Für die beiden baugleichen Analysegeräte **Cobas8000-1** und **Cobas8000-2** wurden die
drei OEE-Faktoren — Verfügbarkeit, Leistungsgrad und Qualität — über einen Zeitraum von
**24 Monaten (März 2024 bis Februar 2026)** einzeln als monatliche Mittelwertlinien
visualisiert. Zusätzlich wurde das berechnete Gesamt-OEE (A×P×Q) dem gespeicherten
Ist-OEE gegenübergestellt.

Vor der Darstellung wurden folgende Datenbereinigungsschritte vorgenommen:

1. **Entfernung von Chargen ohne Produktion (`total_quantity = 0`):** Leerläufe,
   Kurzstarts und reine Stillstandserfassungen werden explizit ausgeschlossen, da sie
   keine OEE-Aussagekraft besitzen und den Mittelwert nach unten verzerren würden.

2. **Entfernung physikalisch unmöglicher OEE-Werte (`oee < 0` oder `oee > 1`):**
   Solche Werte sind Berechnungsfehler im Quellsystem (z. B. fehlerhafte Stammdaten)
   und werden entfernt. OEE = 0 bleibt bewusst erhalten — eine Charge ohne
   Ausbringung ist ein valides Ereignis und gehört zur Analyse.

3. **Kein Quantil-Trim:** Ein pauschaler Ausschluss der untersten und obersten
   Perzentile (z. B. 5 %–95 %) wird bewusst **nicht** angewendet. Die Daten liegen
   qualitätsgeprüft vor. Ein Trim würde echte Extremereignisse wie Störungen oder
   längere Stillstände entfernen — genau jene Ereignisse, die für eine ehrliche
   OEE-Bewertung relevant sind und den tatsächlichen Istzustand abbilden.

4. **Leistungsgrad-Cap bei 100 %:** Bei mehreren Chargen lag der Leistungsgrad über
   100 %. Das ist physikalisch nicht möglich und entsteht, wenn die hinterlegte
   theoretische Ausbringungsrate im Stammdatensystem zu niedrig gepflegt ist.
   Diese Werte wurden auf 100 % begrenzt, nicht entfernt, da der restliche
   OEE-Wert der Charge weiterhin gültig ist.

#### Was sagt das gepunktete OEE (A×P×Q)?

Die lila gepunktete Linie ist das aus den drei Faktoren manuell berechnete OEE.
Sie liegt bei beiden Geräten konsistent über dem roten Ist-OEE. Das bedeutet:
Das Quellsystem berechnet das OEE nach einer konservativeren Methode — möglicherweise
fließen dort auch Zeiten ein, die in den Einzelkomponenten nicht vollständig erfasst
sind (z. B. nicht zugeordnete Stillstände). Für die relative Bewertung der Faktoren
untereinander ist das unerheblich; für eine absolute OEE-Aussage sollte die
Berechnungslogik des Quellsystems dokumentiert und geprüft werden.


```python
import pandas as pd
import warnings
warnings.filterwarnings("ignore")

# ── 7 · Materialanalyse — OEE-Schwankungen nach Materialnummer ───────────────
#
# Fragen:
#   1. Welche Materialien werden am häufigsten produziert? (Volumen)
#   2. Wie stark schwankt der OEE je Material? (σ)
#   3. Gibt es Materialien mit systematisch niedrigem OEE?
#
# Datenbasis: bereinigter Datensatz beider Cobas8000-Linien (aus cobas_data).
# Sortierung: absteigend nach Chargenzahl (Volumen), Top 10 dargestellt.

MATERIAL_COL = "material_number"
TOP_N        = 10

# Beide Linien zusammenführen, Linienzuordnung als Spalte erhalten
df_both = pd.concat(
    [d["df_clean"].assign(linie=label) for label, d in cobas_data.items()],
    ignore_index=True
)

# Chargen ohne Materialnummer ausschließen
n_missing = df_both[MATERIAL_COL].isna().sum()
if n_missing > 0:
    print(f"  {n_missing} Charge(n) ohne Materialnummer ausgeschlossen.\n")
df_both = df_both.dropna(subset=[MATERIAL_COL])

# Materialnummer als sauberer String (führende Nullen erhalten)
df_both[MATERIAL_COL] = df_both[MATERIAL_COL].astype(str).str.strip()

# Anfahrkontrollen ausschließen
EXCLUDE_MATERIALS = ["Anfahrkontrolle"]
n_excluded = df_both[MATERIAL_COL].isin(EXCLUDE_MATERIALS).sum()
if n_excluded > 0:
    print(f"  {n_excluded} Charge(n) als Anfahrkontrolle ausgeschlossen.\n")
df_both = df_both[~df_both[MATERIAL_COL].isin(EXCLUDE_MATERIALS)]

# ── Aggregation je Materialnummer ──────────────────────────────────────────
mat_stats = (
    df_both.groupby(MATERIAL_COL)
    .agg(
        chargen    = ("oee_pct",          "count"),
        oee_mean   = ("oee_pct",          "mean"),
        oee_std    = ("oee_pct",          "std"),
        avail_mean = ("availability_pct", "mean"),
        perf_mean  = ("performance_pct",  "mean"),
        qual_mean  = ("quality_pct",      "mean"),
        qty_total  = ("total_quantity",   "sum"),
    )
    .reset_index()
)

# Einzelchargen-Materialien: σ = NaN → 0
mat_stats["oee_std"] = mat_stats["oee_std"].fillna(0)

# Absteigend nach Chargenzahl
mat_stats = mat_stats.sort_values("chargen", ascending=False).reset_index(drop=True)
mat_top   = mat_stats.head(TOP_N).copy()

# ── Ausgabe ────────────────────────────────────────────────────────────────
overall_both = df_both["oee_pct"].mean()

print(f"Materialanalyse — Cobas8000-1 & Cobas8000-2 kombiniert")
print(f"Gesamt-Ø OEE über alle Chargen:  {overall_both:.2f} %")
print(f"Eindeutige Materialien (bereinigt): {len(mat_stats)}")
print(f"Darstellung: Top {TOP_N} nach Chargenzahl (absteigend)\n")

disp = mat_top[[
    MATERIAL_COL, "chargen", "qty_total",
    "oee_mean", "oee_std",
    "avail_mean", "perf_mean", "qual_mean",
]].copy()

disp.columns = [
    "Material", "Chargen", "Gesamtmenge",
    "Ø OEE", "σ OEE",
    "Ø Verfügb.", "Ø Leistung", "Ø Qualität",
]
for col in disp.columns[3:]:
    disp[col] = disp[col].round(1)

print(disp.to_string(index=False))

# ── Auffälligkeiten ────────────────────────────────────────────────────────
print("\n── Auffälligkeiten (innerhalb Top 10) ───────────────────────────────")

low_thresh = mat_top["oee_mean"].quantile(0.25)
high_std   = mat_top["oee_std"].quantile(0.75)

low_oee = mat_top[mat_top["oee_mean"] <= low_thresh]
if not low_oee.empty:
    print(f"\n  Niedriger Ø OEE (≤ {low_thresh:.1f} %, unteres Quartil):")
    for _, r in low_oee.iterrows():
        print(f"    {r[MATERIAL_COL]:<15}  "
              f"Chargen: {int(r['chargen']):>4}  "
              f"Ø OEE: {r['oee_mean']:.1f} %  "
              f"σ: {r['oee_std']:.1f} %")

vol_high_std = mat_top[mat_top["oee_std"] >= high_std]
if not vol_high_std.empty:
    print(f"\n  Hohe OEE-Schwankung (σ ≥ {high_std:.1f} %, oberes Quartil):")
    for _, r in vol_high_std.iterrows():
        print(f"    {r[MATERIAL_COL]:<15}  "
              f"Chargen: {int(r['chargen']):>4}  "
              f"Ø OEE: {r['oee_mean']:.1f} %  "
              f"σ: {r['oee_std']:.1f} %")
```

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# ── 8 · Materialanalyse — Visualisierung ─────────────────────────────────────
#
# Drei gestapelte Teildiagramme (gemeinsame X-Achse = Materialnummer):
#   ① Chargenzahl je Material   → Produktionsvolumen
#   ② Ø OEE ± σ je Material     → Leistungsniveau & Schwankungsbreite
#   ③ Heatmap OEE-Komponenten   → Wo liegt der OEE-Verlust je Material?
#
# Materialien links = höchstes Volumen (absteigend sortiert).

labels    = mat_top[MATERIAL_COL].tolist()
overall_m = df_both["oee_pct"].mean()

fig = make_subplots(
    rows=3, cols=1,
    shared_xaxes=True,
    vertical_spacing=0.06,
    subplot_titles=(
        f"① Produktionsvolumen — Top {TOP_N} Materialien (Chargenzahl)",
        "② Ø OEE ± 1σ je Material  ·  Fehlerbalken = Standardabweichung",
        "③ OEE-Komponenten je Material (Ø [%])",
    ),
    row_heights=[0.22, 0.42, 0.36],
)

# ── ① Volumen-Balken ──────────────────────────────────────────────────────
# Datenbeschriftungen auf jedem Balken machen den genauen Wert ablesbar,
# da die Balken im engen Bereich (14–19) optisch ähnlich wirken.
# textposition="outside" platziert die Zahl knapp über dem Balken.
chargen_vals = mat_top["chargen"].tolist()
fig.add_trace(
    go.Bar(
        x             = labels,
        y             = chargen_vals,
        name          = "Chargenzahl",
        marker_color  = "#4C72B0",
        opacity       = 0.82,
        text          = chargen_vals,
        textposition  = "outside",
        textfont      = dict(size=11, color="#333333"),
        hovertemplate = (
            "<b>Material %{x}</b><br>"
            "Chargen: %{y}<br>"
            "Gesamtmenge: %{customdata:,.0f}"
            "<extra></extra>"
        ),
        customdata = mat_top["qty_total"].tolist(),
    ),
    row=1, col=1,
)

# ── ② Ø OEE + σ-Fehlerbalken ─────────────────────────────────────────────
# Einheitliche Farbe (#4C72B0) für alle Balken: sauber, neutral und konsistent
# mit den übrigen Diagrammen im Notebook. Ein rot→grün-Gradient wäre redundant,
# da das relative OEE-Niveau bereits durch die Balkenhöhe und die
# σ-Fehlerbalken gut ablesbar ist.
# Die Fehlerbalken (σ) zeigen, wie stark der OEE von Charge zu Charge schwankt:
#   - kurze Fehlerbalken = stabiler, reproduzierbarer Prozess
#   - lange Fehlerbalken = hohe Variabilität, unklare Ursachen
fig.add_trace(
    go.Bar(
        x             = labels,
        y             = mat_top["oee_mean"].round(1).tolist(),
        name          = "Ø OEE",
        marker_color  = "#4C72B0",
        opacity       = 0.85,
        error_y       = dict(
            type      = "data",
            array     = mat_top["oee_std"].round(1).tolist(),
            visible   = True,
            color     = "#C44E52",   # rot → fällt von den blauen Balken ab
            thickness = 1.8,
            width     = 6,
        ),
        hovertemplate = (
            "<b>Material %{x}</b><br>"
            "Ø OEE: %{y:.1f} %<br>"
            "σ (Schwankung): %{customdata:.1f} %"
            "<extra></extra>"
        ),
        customdata = mat_top["oee_std"].round(1).tolist(),
    ),
    row=2, col=1,
)

# Gesamt-Ø Referenzlinie
fig.add_hline(
    y                   = overall_m,
    line                = dict(color="#C44E52", width=1.8, dash="dash"),
    annotation_text     = f"Gesamt-Ø {overall_m:.1f} %",
    annotation_position = "top right",
    annotation_font     = dict(color="#C44E52", size=11),
    row=2, col=1,
)

# ── ③ Heatmap OEE-Komponenten ─────────────────────────────────────────────
comp_matrix = [
    mat_top["avail_mean"].round(1).tolist(),
    mat_top["perf_mean"].round(1).tolist(),
    mat_top["qual_mean"].round(1).tolist(),
]
comp_labels = ["Verfügbarkeit", "Leistungsgrad", "Qualität"]

fig.add_trace(
    go.Heatmap(
        z             = comp_matrix,
        x             = labels,
        y             = comp_labels,
        colorscale    = "RdYlGn",
        zmin          = 0,
        zmax          = 100,
        text          = [[f"{v:.1f}" for v in row] for row in comp_matrix],
        texttemplate  = "%{text}",
        hovertemplate = (
            "<b>Material %{x}</b><br>"
            "%{y}: %{z:.1f} %"
            "<extra></extra>"
        ),
        showscale = True,
        colorbar  = dict(
            title     = "%",
            thickness = 12,
            len       = 0.30,
            y         = 0.09,
            yanchor   = "bottom",
        ),
    ),
    row=3, col=1,
)

# ── Layout ────────────────────────────────────────────────────────────────
fig.update_layout(
    title = dict(
        text = (
            f"Materialanalyse — OEE-Schwankungen nach Materialnummer<br>"
            f"<sup>Top {TOP_N} Materialien nach Chargenzahl (absteigend)  ·  "
            f"Cobas8000-1 & Cobas8000-2  ·  "
            f"Gesamt-Ø OEE: {overall_m:.1f} %  ·  "
            f"{len(mat_stats)} Materialien gesamt</sup>"
        ),
        font = dict(size=16),
    ),
    height        = 980,
    width         = 1100,
    plot_bgcolor  = "white",
    paper_bgcolor = "#f8f8f8",
    showlegend    = False,
    hovermode     = "x unified",
    bargap        = 0.22,
)

# Y-Achse Panel ①: startet knapp unterhalb des Minimums statt bei 0,
# damit Unterschiede zwischen den Balken (14–19 Chargen) klar sichtbar werden.
# clip setzt einen kleinen Puffer nach unten und oben.
chargen_min = min(chargen_vals)
chargen_max = max(chargen_vals)
fig.update_yaxes(title_text="Chargen",
                 range=[chargen_min - 2, chargen_max + 3],
                 showgrid=True, gridcolor="#e8e8e8", row=1, col=1)
fig.update_yaxes(title_text="OEE [%]",  range=[0, 100],
                 showgrid=True, gridcolor="#e8e8e8", row=2, col=1)
fig.update_yaxes(title_text="",         row=3, col=1)
fig.update_xaxes(title_text="Materialnummer",
                 tickangle=-40, tickfont=dict(size=10), row=3, col=1)

fig.show()
```
